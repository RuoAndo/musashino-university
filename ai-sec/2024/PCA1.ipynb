```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# 1. データセットの読み込み
data = load_iris()
X = data.data  # 特徴量
y = data.target  # ラベル

# 2. データの標準化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. PCAモデルの構築と適用
pca = PCA(n_components=2)  # 2次元に次元削減
X_pca = pca.fit_transform(X_scaled)

# 4. 主成分の寄与率を表示
explained_variance_ratio = pca.explained_variance_ratio_
print(f"主成分1の寄与率: {explained_variance_ratio[0]:.2f}")
print(f"主成分2の寄与率: {explained_variance_ratio[1]:.2f}")

# 5. 次元削減後のデータを可視化
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=50, edgecolor='k')
plt.title("PCA: Iris Dataset Projected to 2D")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(scatter, label="Class Label")
plt.grid(True)
plt.show()
```
