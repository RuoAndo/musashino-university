```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# 1. 
data = load_iris()
X = data.data  # 
y = data.target  # 

# 2. 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. 3
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_scaled[:, 0], X_scaled[:, 1], X_scaled[:, 2], c=y, cmap='viridis', s=50, edgecolor='k')

# 
ax.set_title("3D Scatter Plot of Scaled Data (Before PCA)", fontsize=15)
ax.set_xlabel("Feature 1 (scaled)")
ax.set_ylabel("Feature 2 (scaled)")
ax.set_zlabel("Feature 3 (scaled)")

# 
plt.colorbar(scatter, ax=ax, label="Class Label")
plt.show()

# 4. PCA
pca = PCA(n_components=2)  # 2
X_pca = pca.fit_transform(X_scaled)

# 5. 
explained_variance_ratio = pca.explained_variance_ratio_
print(f"1: {explained_variance_ratio[0]:.2f}")
print(f"2: {explained_variance_ratio[1]:.2f}")

# 6. 2
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=50, edgecolor='k')
plt.title("2D Scatter Plot After PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(scatter, label="Class Label")
plt.grid(True)
plt.show()
```
