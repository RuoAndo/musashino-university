```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# 1. データセットの読み込み
data = load_iris()
X = data.data  # 特徴量
y = data.target  # ラベル

# 2. データの標準化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. 次元削減前の3次元プロット
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_scaled[:, 0], X_scaled[:, 1], X_scaled[:, 2], c=y, cmap='viridis', s=50, edgecolor='k')

# 軸ラベルの設定
ax.set_title("3D Scatter Plot of Scaled Data (Before PCA)", fontsize=15)
ax.set_xlabel("Feature 1 (scaled)")
ax.set_ylabel("Feature 2 (scaled)")
ax.set_zlabel("Feature 3 (scaled)")

# カラーバーの追加
plt.colorbar(scatter, ax=ax, label="Class Label")
plt.show()

# 4. PCAモデルの構築と適用
pca = PCA(n_components=2)  # 2次元に次元削減
X_pca = pca.fit_transform(X_scaled)

# 5. 主成分の寄与率を表示
explained_variance_ratio = pca.explained_variance_ratio_
print(f"主成分1の寄与率: {explained_variance_ratio[0]:.2f}")
print(f"主成分2の寄与率: {explained_variance_ratio[1]:.2f}")

# 6. 次元削減後の2次元プロット
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=50, edgecolor='k')
plt.title("2D Scatter Plot After PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.colorbar(scatter, label="Class Label")
plt.grid(True)
plt.show()
```
