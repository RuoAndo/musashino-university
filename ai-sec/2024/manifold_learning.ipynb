```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import TSNE, LocallyLinearEmbedding

# 1. スイスロールデータセットの生成
n_samples = 1000
X, color = make_swiss_roll(n_samples=n_samples, noise=0.1, random_state=42)

# スイスロールの3Dプロット
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap='Spectral', s=50)
ax.set_title("Swiss Roll Dataset (3D)")
plt.colorbar(scatter, label="Color")
plt.show()

# 2. t-SNEによる次元削減
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
X_tsne = tsne.fit_transform(X)

# t-SNEの結果をプロット
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=color, cmap='Spectral', s=50)
plt.title("t-SNE: Swiss Roll in 2D")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(label="Color")
plt.grid(True)
plt.show()

# 3. LLEによる次元削減
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=12, random_state=42)
X_lle = lle.fit_transform(X)

# LLEの結果をプロット
plt.figure(figsize=(8, 6))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=color, cmap='Spectral', s=50)
plt.title("LLE: Swiss Roll in 2D")
plt.xlabel("LLE Component 1")
plt.ylabel("LLE Component 2")
plt.colorbar(label="Color")
plt.grid(True)
plt.show()
```
