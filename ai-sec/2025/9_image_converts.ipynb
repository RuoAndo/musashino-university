```python
# ===========================================
#  Xception 軽量版・完全スクリプト：
#  - 猫画像の中間層可視化（層を一部に限定＋画像小さめ）
#  - Xception フィルタ可視化（画像・フィルタ数を縮小）
#  - 象画像に対する Grad-CAM（画像小さめ）
# ===========================================

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.cm as cm

# -------------------------------
# 0. Xception（畳み込みベース）読み込み
# -------------------------------
# 中間層とフィルタ可視化に使うモデル（トップ層なし）
base_model = keras.applications.xception.Xception(
    weights="imagenet",
    include_top=False
)
print("Loaded Xception base (include_top=False)")
base_model.summary()

# -------------------------------------------------
# 1. 猫画像を取得して、Xception 用に前処理
# -------------------------------------------------

# cat.jpg を自動取得（初回のみ、それ以降はキャッシュ）
cat_img_path = keras.utils.get_file(
    fname="cat.jpg",
    origin="https://img-datasets.s3.amazonaws.com/cat.jpg"
)

def get_img_array_for_xception(img_path, target_size=(299, 299)):
    """Xception 用の前処理込みで (1, h, w, 3) のテンソルを返す"""
    img = keras.utils.load_img(img_path, target_size=target_size)
    array = keras.utils.img_to_array(img)
    array = np.expand_dims(array, axis=0)
    array = keras.applications.xception.preprocess_input(array)
    return array

# 表示用に、前処理前の画像も別に読み込んでおく
def load_img_raw(img_path, target_size=(299, 299)):
    img = keras.utils.load_img(img_path, target_size=target_size)
    return keras.utils.img_to_array(img)

cat_img_raw = load_img_raw(cat_img_path, target_size=(299, 299))
cat_img_batch = get_img_array_for_xception(cat_img_path, target_size=(299, 299))

plt.figure(figsize=(4, 4))  # 小さめの図
plt.axis("off")
plt.title("Input cat image")
plt.imshow(cat_img_raw.astype("uint8"))
plt.show()

# -------------------------------------------------
# 2. 中間層（Conv / SepConv / MaxPool）の活性可視化（軽量版）
# -------------------------------------------------

layer_outputs = []
layer_names = []

for layer in base_model.layers:
    if isinstance(layer, (layers.Conv2D,
                          layers.SeparableConv2D,
                          layers.MaxPooling2D)):
        layer_outputs.append(layer.output)
        layer_names.append(layer.name)

# ★ 可視化する層の数を制限（先頭から 8 層だけ）
max_layers_to_show = 8
layer_outputs = layer_outputs[:max_layers_to_show]
layer_names = layer_names[:max_layers_to_show]

activation_model = keras.Model(
    inputs=base_model.input,
    outputs=layer_outputs
)

# 1 枚の猫画像を通して各層の出力を取得
activations = activation_model.predict(cat_img_batch)
print("Num of selected layers:", len(activations))

# 例：最初に選ばれた層の出力形状
first_layer_activation = activations[0]
print("First selected layer activation shape:", first_layer_activation.shape)

# 例：その層の 5 番目のチャネルを表示（画像小さめ）
plt.figure(figsize=(3, 3))
plt.title(f"{layer_names[0]} - channel 5")
plt.axis("off")
plt.imshow(first_layer_activation[0, :, :, 5], cmap="viridis")
plt.show()

# ---- 全中間層のチャネルをタイル状に可視化（軽量仕様） ----

images_per_row = 8  # 行あたりのチャネル数を減らす

for layer_name, layer_activation in zip(layer_names, activations):
    # layer_activation.shape: (1, size, size, n_features)
    n_features = layer_activation.shape[-1]  # チャネル数
    size = layer_activation.shape[1]        # 高さ = 幅

    n_cols = n_features // images_per_row
    if n_cols == 0:
        # チャネル数が少なすぎる層はスキップ
        continue

    display_grid = np.zeros(
        ((size + 1) * n_cols - 1, (size + 1) * images_per_row - 1)
    )

    for col in range(n_cols):
        for row in range(images_per_row):
            channel_index = col * images_per_row + row
            if channel_index >= n_features:
                break

            channel_image = layer_activation[0, :, :, channel_index].copy()

            if channel_image.sum() != 0:
                channel_image -= channel_image.mean()
                channel_image /= (channel_image.std() + 1e-5)
                channel_image *= 64
                channel_image += 128

            channel_image = np.clip(channel_image, 0, 255).astype("uint8")

            v_start = col * (size + 1)
            v_end = v_start + size
            h_start = row * (size + 1)
            h_end = h_start + size
            display_grid[v_start:v_end, h_start:h_end] = channel_image

    # ★ 常に小さめのサイズで表示
    plt.figure(figsize=(4, 4))
    plt.title(layer_name)
    plt.axis("off")
    plt.imshow(display_grid, aspect="auto", cmap="viridis")
    plt.show()

# -------------------------------------------------
# 3. フィルタ可視化（勾配上昇で「好むパターン」を生成：軽量版）
# -------------------------------------------------

# 可視化対象の層を選ぶ（本と同じく block3_sepconv1 を使用）
layer_name = "block3_sepconv1"
target_layer = base_model.get_layer(name=layer_name)

feature_extractor = keras.Model(
    inputs=base_model.input,
    outputs=target_layer.output
)

def compute_loss(image, filter_index):
    """指定フィルタの活性平均を損失として計算"""
    activation = feature_extractor(image)
    filter_activation = activation[:, 2:-2, 2:-2, filter_index]  # 端はカット
    return tf.reduce_mean(filter_activation)

@tf.function
def gradient_ascent_step(image, filter_index, learning_rate=10.0):
    with tf.GradientTape() as tape:
        tape.watch(image)
        loss = compute_loss(image, filter_index)
    grads = tape.gradient(loss, image)
    grads = tf.math.l2_normalize(grads)
    image = image + learning_rate * grads
    return image

# ★ 画像サイズを小さく
img_width = 128
img_height = 128

def generate_filter_pattern(filter_index,
                            iterations=30,
                            learning_rate=10.0):
    """指定フィルタを最大にする画像を生成"""
    image = tf.random.uniform(
        shape=(1, img_width, img_height, 3),
        minval=0.4,
        maxval=0.6
    )
    for i in range(iterations):
        image = gradient_ascent_step(
            image, filter_index=filter_index, learning_rate=learning_rate
        )
    return image[0].numpy()

def deprocess_image(image):
    """生成画像を可視化用に正規化＆クリップ"""
    image -= image.mean()
    image /= (image.std() + 1e-5)
    image *= 64
    image += 128
    image = np.clip(image, 0, 255).astype("uint8")
    # 周辺 16 ピクセルをカットして中央だけ残す（カット幅も少し縮小）
    margin_cut = 16
    image = image[margin_cut:-margin_cut, margin_cut:-margin_cut, :]
    return image

# 例：フィルタ 2 のパターンを一枚表示（小さめ）
pattern = deprocess_image(generate_filter_pattern(filter_index=2))
plt.figure(figsize=(3, 3))
plt.axis("off")
plt.title(f"Filter 2 pattern in {layer_name}")
plt.imshow(pattern)
plt.show()

# ★ 生成するフィルタ数を 64 → 16 個（4x4）に削減
num_filters = 16
all_images = []
for filter_index in range(num_filters):
    print(f"Processing filter {filter_index}")
    img = deprocess_image(generate_filter_pattern(filter_index))
    all_images.append(img)

margin = 4
n = 4  # 4x4 = 16 フィルタ

cropped_width = all_images[0].shape[1]
cropped_height = all_images[0].shape[0]
total_width = n * cropped_width + (n - 1) * margin
total_height = n * cropped_height + (n - 1) * margin
stitched_filters = np.zeros((total_height, total_width, 3), dtype="uint8")

for i in range(n):
    for j in range(n):
        image = all_images[i * n + j]
        v_start = i * (cropped_height + margin)
        v_end = v_start + cropped_height
        h_start = j * (cropped_width + margin)
        h_end = h_start + cropped_width
        stitched_filters[v_start:v_end, h_start:h_end, :] = image

filters_png_name = f"filters_for_layer_{layer_name}_compact.png"
keras.utils.save_img(filters_png_name, stitched_filters)
print("Saved filter grid to:", filters_png_name)

# ★ 表示も小さめ
plt.figure(figsize=(4, 4))
plt.axis("off")
plt.title(f"Filters for layer {layer_name} (compact)")
plt.imshow(stitched_filters)
plt.show()

# -------------------------------------------------
# 4. Grad-CAM：象画像に対して「どこを見ているか」可視化（軽量版）
# -------------------------------------------------

# Grad-CAM 用にトップ層つき Xception を読み込む
full_model = keras.applications.xception.Xception(weights="imagenet")
print("Loaded full Xception (include_top=True)")

def get_img_array_for_xception_full(img_path, target_size=(299, 299)):
    img = keras.utils.load_img(img_path, target_size=target_size)
    array = keras.utils.img_to_array(img)
    array = np.expand_dims(array, axis=0)
    array = keras.applications.xception.preprocess_input(array)
    return array

# 象画像を取得
elephant_img_path = keras.utils.get_file(
    fname="elephant.jpg",
    origin="https://img-datasets.s3.amazonaws.com/elephant.jpg"
)

elephant_img_array = get_img_array_for_xception_full(
    elephant_img_path, target_size=(299, 299)
)

# 予測と上位3クラス
preds = full_model.predict(elephant_img_array)
top3 = keras.applications.xception.decode_predictions(preds, top=3)[0]
print("Top-3 predictions for elephant.jpg:")
for cls_id, name, score in top3:
    print(f"{cls_id}: {name} ({score:.3f})")

top_pred_index = np.argmax(preds[0])

# Grad-CAM で使う最後の畳み込み層
last_conv_layer_name = "block14_sepconv2_act"
last_conv_layer = full_model.get_layer(last_conv_layer_name)

last_conv_layer_model = keras.Model(
    full_model.inputs, last_conv_layer.output
)

classifier_layer_names = [
    "avg_pool",
    "predictions",
]

classifier_input = keras.Input(shape=last_conv_layer.output.shape[1:])
x = classifier_input
for name in classifier_layer_names:
    x = full_model.get_layer(name)(x)
classifier_model = keras.Model(classifier_input, x)

# 勾配計算
with tf.GradientTape() as tape:
    last_conv_layer_output = last_conv_layer_model(elephant_img_array)
    tape.watch(last_conv_layer_output)
    preds_for_grad = classifier_model(last_conv_layer_output)
    top_class_channel = preds_for_grad[:, top_pred_index]

grads = tape.gradient(top_class_channel, last_conv_layer_output)

# チャネルごとの勾配平均（重要度）
pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2)).numpy()
last_conv_layer_output = last_conv_layer_output.numpy()[0]

for i in range(pooled_grads.shape[-1]):
    last_conv_layer_output[:, :, i] *= pooled_grads[i]

# チャネル方向に平均してヒートマップ作成
heatmap = np.mean(last_conv_layer_output, axis=-1)
heatmap = np.maximum(heatmap, 0)
heatmap /= (heatmap.max() + 1e-8)

plt.figure(figsize=(3, 3))
plt.title("Grad-CAM heatmap")
plt.axis("off")
plt.imshow(heatmap, cmap="viridis")
plt.show()

# ヒートマップを元画像に重ねる
orig_elephant_img = keras.utils.load_img(elephant_img_path)
orig_elephant_img = keras.utils.img_to_array(orig_elephant_img)

heatmap_uint8 = np.uint8(255 * heatmap)

jet = cm.get_cmap("jet")
jet_colors = jet(np.arange(256))[:, :3]
jet_heatmap = jet_colors[heatmap_uint8]

jet_heatmap = keras.utils.array_to_img(jet_heatmap)
jet_heatmap = jet_heatmap.resize(
    (orig_elephant_img.shape[1], orig_elephant_img.shape[0])
)
jet_heatmap = keras.utils.img_to_array(jet_heatmap)

superimposed_img = jet_heatmap * 0.4 + orig_elephant_img
superimposed_img = keras.utils.array_to_img(superimposed_img)

cam_save_path = "elephant_cam_compact.jpg"
superimposed_img.save(cam_save_path)
print("Saved Grad-CAM superimposed image to:", cam_save_path)

plt.figure(figsize=(4, 4))  # 小さめの表示
plt.axis("off")
plt.title("Grad-CAM on elephant.jpg (compact)")
plt.imshow(superimposed_img)
plt.show()
```
