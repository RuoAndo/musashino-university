```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# 観測データ（1=当たり、0=はずれ）
data = [1, 0, 0, 1, 0]   # 当たり2, はずれ3

# --- 事前分布 ---
alpha_prior = 1   # Beta(1,1) = 一様事前
beta_prior  = 1

# --- 観測データをカウント ---
k = sum(data)        # 当たり回数
n = len(data)        # 試行回数

# --- 事後分布 ---
alpha_post = alpha_prior + k
beta_post  = beta_prior  + (n - k)

print("事後分布 = Beta({}, {})".format(alpha_post, beta_post))
print("事後平均 =", alpha_post / (alpha_post + beta_post))
print("次の1回が当たる確率 =", alpha_post / (alpha_post + beta_post))

# --- 事後分布を描画 ---
x = np.linspace(0, 1, 400)
y = beta.pdf(x, alpha_post, beta_post)

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Posterior Beta({}, {})".format(alpha_post, beta_post))
plt.xlabel("p (当たり確率)")
plt.ylabel("density")
plt.grid()
plt.show()
```
