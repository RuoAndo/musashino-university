```python
# -*- coding: utf-8 -*-

import datetime
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt


# =====================================================
# 設定
# =====================================================

n_observations = 100

stocks = ["AAPL", "GOOG", "TSLA", "AMZN"]

startdate = "2012-09-01"
enddate = "2015-04-27"


# =====================================================
# 株価データ取得
# =====================================================

stock_closes = pd.DataFrame()

for stock in stocks:
    print(f"{stock} のデータを取得中...")

    df = yf.download(
        stock,
        start=startdate,
        end=enddate,
        progress=False,
        auto_adjust=False
    )

    if df.empty:
        print(f"{stock} のデータが取得できませんでした。")
        continue

    stock_closes[stock] = df["Close"]


# =====================================================
# 日付順に整理
# =====================================================

stock_closes = stock_closes.sort_index()

print("\n終値データ:")
print(stock_closes.head())
print(stock_closes.tail())


# =====================================================
# 日次リターン計算
# =====================================================

stock_returns = stock_closes.pct_change().dropna()

stock_returns = stock_returns.tail(n_observations)

dates = stock_returns.index.to_list()


print("\n直近100日分の日次リターン:")
print(stock_returns.head())
print(stock_returns.tail())

print("\n日付:")
print(dates[:5])
print("...")
print(dates[-5:])


# =====================================================
# 終値のプロット
# =====================================================

plt.figure(figsize=(12, 6))

for stock in stocks:
    if stock in stock_closes.columns:
        plt.plot(stock_closes.index, stock_closes[stock], label=stock)

plt.title("Stock Close Prices")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# =====================================================
# 日次リターンのプロット
# =====================================================

plt.figure(figsize=(12, 6))

for stock in stocks:
    if stock in stock_returns.columns:
        plt.plot(stock_returns.index, stock_returns[stock], label=stock)

plt.title("Daily Returns - Most Recent 100 Observations")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# =====================================================
# リターンの相関行列
# =====================================================

corr = stock_returns.corr()

print("\nリターンの相関行列:")
print(corr)


# =====================================================
# 相関行列のヒートマップ
# =====================================================

plt.figure(figsize=(6, 5))
plt.imshow(corr, interpolation="nearest")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.index)), corr.index)

plt.title("Correlation Matrix of Stock Returns")
plt.tight_layout()
plt.show()

# -*- coding: utf-8 -*-

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import yfinance as yf
import pymc as pm


# =====================================================
# 1. 設定
# =====================================================

stocks = ["AAPL", "GOOG", "TSLA", "AMZN"]

startdate = "2012-09-01"
enddate = "2015-04-27"

n_observations = 100

random_seed = 42


# =====================================================
# 2. 株価データ取得
# =====================================================

stock_closes = pd.DataFrame()

for stock in stocks:
    print(f"{stock} のデータを取得中...")

    df = yf.download(
        stock,
        start=startdate,
        end=enddate,
        progress=False,
        auto_adjust=False
    )

    if df.empty:
        raise ValueError(f"{stock} のデータが取得できませんでした。")

    stock_closes[stock] = df["Close"]

stock_closes = stock_closes.sort_index()

print("\n終値データ:")
print(stock_closes.head())
print(stock_closes.tail())


# =====================================================
# 3. 日次リターン計算
# =====================================================

stock_returns = stock_closes.pct_change().dropna()
stock_returns = stock_returns.tail(n_observations)

returns_data = stock_returns.values

n_assets = len(stocks)

print("\n日次リターン:")
print(stock_returns.head())
print(stock_returns.tail())

print("\nreturns_data shape:", returns_data.shape)


# =====================================================
# 4. 事前分布
# =====================================================

expert_prior_params = {
    "AAPL": (0.0010, 0.020),
    "GOOG": (0.0010, 0.020),
    "TSLA": (0.0015, 0.035),
    "AMZN": (0.0010, 0.025),
}

prior_mu = np.array([expert_prior_params[s][0] for s in stocks])
prior_std = np.array([expert_prior_params[s][1] for s in stocks])

print("\n事前平均:")
print(prior_mu)

print("\n事前標準偏差:")
print(prior_std)


# =====================================================
# 5. PyMC モデル
# =====================================================

with pm.Model() as model:

    # 平均リターン
    mu = pm.Normal(
        "mu",
        mu=prior_mu,
        sigma=prior_std,
        shape=n_assets
    )

    # 共分散構造
    chol, corr, sigmas = pm.LKJCholeskyCov(
        "chol_cov",
        n=n_assets,
        eta=2.0,
        sd_dist=pm.Exponential.dist(1.0),
        compute_corr=True
    )

    # 観測データ
    obs = pm.MvNormal(
        "obs",
        mu=mu,
        chol=chol,
        observed=returns_data
    )

    trace = pm.sample(
        draws=2000,
        tune=1000,
        chains=2,
        cores=1,
        target_accept=0.9,
        random_seed=random_seed
    )


# =====================================================
# 6. posterior からサンプルを取り出す
# =====================================================

mu_samples = trace.posterior["mu"].values
sigma_samples = trace.posterior["chol_cov_stds"].values
corr_samples = trace.posterior["chol_cov_corr"].values

# shape:
# mu_samples:    chain, draw, asset
# sigma_samples: chain, draw, asset
# corr_samples:  chain, draw, asset, asset

mu_flat = mu_samples.reshape(-1, n_assets)
sigma_flat = sigma_samples.reshape(-1, n_assets)
corr_flat = corr_samples.reshape(-1, n_assets, n_assets)

print("\nmu_samples shape:", mu_samples.shape)
print("sigma_samples shape:", sigma_samples.shape)
print("corr_samples shape:", corr_samples.shape)


# =====================================================
# 7. 事後平均・信用区間の表示
# =====================================================

summary_rows = []

for i, stock in enumerate(stocks):
    samples = mu_flat[:, i]

    summary_rows.append({
        "stock": stock,
        "posterior_mean_return": samples.mean(),
        "posterior_sd_return": samples.std(),
        "hdi_2.5%": np.quantile(samples, 0.025),
        "hdi_97.5%": np.quantile(samples, 0.975),
    })

summary_df = pd.DataFrame(summary_rows)

print("\n平均リターン mu の事後要約:")
print(summary_df)


# =====================================================
# 8. 平均リターン mu の事後分布
# =====================================================

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, stock in enumerate(stocks):
    samples = mu_flat[:, i]

    axes[i].hist(
        samples,
        bins=40,
        density=True
    )

    axes[i].axvline(
        samples.mean(),
        linestyle="--",
        linewidth=2
    )

    axes[i].set_title(f"{stock} posterior mu")
    axes[i].set_xlabel("daily return")
    axes[i].set_ylabel("density")

plt.tight_layout()
plt.show()


# =====================================================
# 9. 標準偏差 sigma の事後分布
# =====================================================

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, stock in enumerate(stocks):
    samples = sigma_flat[:, i]

    axes[i].hist(
        samples,
        bins=40,
        density=True
    )

    axes[i].axvline(
        samples.mean(),
        linestyle="--",
        linewidth=2
    )

    axes[i].set_title(f"{stock} posterior sigma")
    axes[i].set_xlabel("daily volatility")
    axes[i].set_ylabel("density")

plt.tight_layout()
plt.show()


# =====================================================
# 10. 事後平均の相関行列
# =====================================================

corr_mean = corr_flat.mean(axis=0)

corr_df = pd.DataFrame(
    corr_mean,
    index=stocks,
    columns=stocks
)

print("\n事後平均の相関行列:")
print(corr_df)

plt.figure(figsize=(6, 5))
plt.imshow(corr_df, interpolation="nearest")
plt.colorbar()

plt.xticks(range(n_assets), stocks)
plt.yticks(range(n_assets), stocks)

plt.title("Posterior Mean Correlation Matrix")
plt.tight_layout()
plt.show()


# =====================================================
# 11. 共分散行列の事後平均
# =====================================================

cov_samples = []

for k in range(corr_flat.shape[0]):
    s = sigma_flat[k]
    c = corr_flat[k]

    cov = np.diag(s) @ c @ np.diag(s)
    cov_samples.append(cov)

cov_samples = np.array(cov_samples)
cov_mean = cov_samples.mean(axis=0)

cov_df = pd.DataFrame(
    cov_mean,
    index=stocks,
    columns=stocks
)

print("\n事後平均の共分散行列:")
print(cov_df)


# =====================================================
# 12. リスク・リターン散布図
# =====================================================

posterior_mean_return = mu_flat.mean(axis=0)
posterior_mean_volatility = sigma_flat.mean(axis=0)

risk_return_df = pd.DataFrame({
    "stock": stocks,
    "posterior_mean_return": posterior_mean_return,
    "posterior_mean_volatility": posterior_mean_volatility
})

print("\nリスク・リターン:")
print(risk_return_df)

plt.figure(figsize=(8, 6))

plt.scatter(
    posterior_mean_volatility,
    posterior_mean_return
)

for i, stock in enumerate(stocks):
    plt.text(
        posterior_mean_volatility[i],
        posterior_mean_return[i],
        stock
    )

plt.xlabel("posterior mean volatility")
plt.ylabel("posterior mean daily return")
plt.title("Bayesian Risk-Return Plot")
plt.grid(True)
plt.tight_layout()
plt.show()


# =====================================================
# 13. 実データの終値プロット
# =====================================================

plt.figure(figsize=(12, 6))

for stock in stocks:
    plt.plot(
        stock_closes.index,
        stock_closes[stock],
        label=stock
    )

plt.title("Stock Close Prices")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# =====================================================
# 14. 実データの日次リターンプロット
# =====================================================

plt.figure(figsize=(12, 6))

for stock in stocks:
    plt.plot(
        stock_returns.index,
        stock_returns[stock],
        label=stock
    )

plt.title("Daily Returns")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


from IPython.core.pylabtools import figsize
colors = ["C0", "C1", "C2", "C3"]

figsize(11., 5 )

for i, _stock in enumerate(stocks):
    plt.subplot(2,2,i+1)
    plt.hist(stock_returns[_stock], bins=20,
             density = True, histtype="stepfilled",
             color=colors[i], alpha=0.7)
    plt.title(_stock + " returns")
    plt.xlim(-0.15, 0.15)

plt.tight_layout()
plt.suptitle("Histogram of daily returns", size =14);

with pm.Model() as model:

    mu = pm.Normal(
        "mu",
        mu=prior_mu,
        sigma=prior_std,
        shape=4
    )

    chol, corr, sigmas = pm.LKJCholeskyCov(
        "chol_cov",
        n=4,
        eta=2,
        sd_dist=pm.Exponential.dist(1.0),
        compute_corr=True
    )

    obs = pm.MvNormal(
        "obs",
        mu=mu,
        chol=chol,           # ← covではなくchol
        observed=stock_returns.values
    )

    trace = pm.sample(
        draws=5000,
        tune=1000,
        chains=1,
        target_accept=0.9
    )
    
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))

# 平均リターン mu の事後サンプル
mu_samples = trace.posterior["mu"].values[0]
# shape: draw, asset

for i in range(4):
    plt.hist(
        mu_samples[:, i],
        alpha=0.8 - 0.05 * i,
        bins=30,
        histtype="stepfilled",
        density=True,
        label=stock_returns.columns[i]
    )

plt.legend()
plt.title("Posterior Distribution of Mean Returns")
plt.xlabel("daily return")
plt.ylabel("density")
plt.tight_layout()
plt.show()


figsize(15.0,8)
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.hist(mu_samples[:,i], alpha = 0.8 - 0.05*i, bins = 30,
             histtype="stepfilled", density=True, color = colors[i],
             label = "%s" % stock_returns.columns[i])
    plt.title("%s" % stock_returns.columns[i])
    plt.xlim(-0.15, 0.15)
    
plt.suptitle("Posterior distribution of daily stock returns")
plt.tight_layout()

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 相関行列と標準偏差の事後サンプルを取得
corr_samples = trace.posterior["chol_cov_corr"].values
std_samples = trace.posterior["chol_cov_stds"].values

# chain, draw をまとめる
corr_samples = corr_samples.reshape(-1, corr_samples.shape[-2], corr_samples.shape[-1])
std_samples = std_samples.reshape(-1, std_samples.shape[-1])

# 共分散行列を復元
cov_samples = []

for k in range(corr_samples.shape[0]):
    D = np.diag(std_samples[k])
    cov = D @ corr_samples[k] @ D
    cov_samples.append(cov)

cov_samples = np.array(cov_samples)

# 事後平均の共分散行列
mean_covariance_matrix = cov_samples.mean(axis=0)

cov_df = pd.DataFrame(
    mean_covariance_matrix,
    index=stock_returns.columns,
    columns=stock_returns.columns
)

print("事後平均の共分散行列:")
print(cov_df)

# 相関行列
mean_correlation_matrix = corr_samples.mean(axis=0)

corr_df = pd.DataFrame(
    mean_correlation_matrix,
    index=stock_returns.columns,
    columns=stock_returns.columns
)

print("\n事後平均の相関行列:")
print(corr_df)

# 相関行列のプロット
plt.figure(figsize=(6, 5))
plt.imshow(mean_correlation_matrix)
plt.colorbar()

plt.xticks(range(len(stock_returns.columns)), stock_returns.columns)
plt.yticks(range(len(stock_returns.columns)), stock_returns.columns)

plt.title("Posterior Mean Correlation Matrix")
plt.tight_layout()
plt.show()

```
