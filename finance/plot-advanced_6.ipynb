```python
# =============================================================================
# BOCPD確率 + 全手法変化点pairs + 元価格時系列 プロット完全版
# =============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path
from datetime import datetime
import time
import re

# =============================================================================
# パラメータ
# =============================================================================

CHANGE_POINT_DIR = Path(r"D:\musashino-university\finance\change_point_output_advanced")
COINGECKO_DIR = Path(r"D:\musashino-university\finance\coingecko_by_coin")

BOCPD_PATTERN = "advanced_change_point_bocpd_probabilities_*.csv"
PAIRS_PATTERN = "advanced_change_point_pairs_ALL_ADVANCED_METHODS_*.csv"

OUTPUT_DIR = CHANGE_POINT_DIR / "plot_bocpd_vs_pairs"
OUTPUT_DIR.mkdir(exist_ok=True)

RUN_ID = datetime.now().strftime("%Y%m%d_%H%M%S")

PLOT_TOP_N = 10
BOCPD_THRESHOLD = 0.8

SHOW_PLOTS = True
SAVE_PLOTS = True
PLOT_DPI = 200

USE_LOG_PRICE = False
NORMALIZE_PRICE = False

MAX_POINTS_PER_SERIES = 3000

ENCODING_OUT = "utf-8-sig"

# =============================================================================
# ログ
# =============================================================================

START_TIME = time.time()
LOG_FILE = OUTPUT_DIR / f"bocpd_vs_pairs_plot_log_{RUN_ID}.txt"

def log(msg=""):
    elapsed = time.time() - START_TIME
    line = f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')} | +{elapsed:.1f}s] {msg}"
    print(line)
    with open(LOG_FILE, "a", encoding="utf-8") as f:
        f.write(line + "\n")

# =============================================================================
# 補助関数
# =============================================================================

def find_column(df, candidates):
    for c in candidates:
        if c in df.columns:
            return c
    return None


def safe_name(s):
    s = str(s)
    s = re.sub(r'[\\/:*?"<>|]', "_", s)
    s = s.replace(" ", "_")
    return s[:120]


def normalize_key(s):
    s = str(s).lower()
    s = Path(s).stem
    s = s.replace(".csv", "")
    s = re.sub(r"[^a-z0-9]+", "", s)
    return s


def downsample_df(df, max_points):
    if len(df) <= max_points:
        return df
    idx = np.linspace(0, len(df) - 1, max_points).astype(int)
    return df.iloc[idx].copy()


def load_latest_file(directory, pattern):
    files = sorted(directory.glob(pattern))
    if not files:
        raise FileNotFoundError(f"対象ファイルが見つかりません: {directory / pattern}")
    return max(files, key=lambda p: p.stat().st_mtime)


def build_file_index(csv_files):
    index = {}
    for p in csv_files:
        for key in {normalize_key(p.name), normalize_key(p.stem)}:
            if key and key not in index:
                index[key] = p
    return index


def find_timeseries_file(series_name, csv_files, file_index):
    s = str(series_name)
    candidates = [s, s + ".csv"]

    p = Path(s)
    if p.suffix.lower() == ".csv":
        candidates += [p.name, p.stem]

    for c in candidates:
        key = normalize_key(c)
        if key in file_index:
            return file_index[key]

    key = normalize_key(s)

    for p in csv_files:
        pk = normalize_key(p.name)

        if key and key in pk:
            return p

        if pk and pk in key:
            return p

    return None


def load_price_csv(csv_path):
    df = pd.read_csv(csv_path)

    datetime_col = find_column(
        df,
        [
            "timestamp",
            "datetime",
            "date",
            "time",
            "Date",
            "Datetime",
            "Timestamp",
            "snapped_at",
        ],
    )

    price_col = find_column(
        df,
        [
            "price",
            "Price",
            "close",
            "Close",
            "market_price",
            "current_price",
            "usd",
            "value",
        ],
    )

    if datetime_col is None:
        object_cols = df.select_dtypes(include=["object"]).columns.tolist()
        for c in object_cols:
            test = pd.to_datetime(df[c], errors="coerce")
            if test.notna().sum() > len(df) * 0.5:
                datetime_col = c
                break

    if price_col is None:
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        if numeric_cols:
            price_col = numeric_cols[0]

    if datetime_col is None:
        raise ValueError(f"日時列が見つかりません: {csv_path.name}")

    if price_col is None:
        raise ValueError(f"価格列が見つかりません: {csv_path.name}")

    df[datetime_col] = pd.to_datetime(df[datetime_col], errors="coerce")
    df[price_col] = pd.to_numeric(df[price_col], errors="coerce")

    df = df[[datetime_col, price_col]].dropna()
    df = df.sort_values(datetime_col)

    return df, datetime_col, price_col


def infer_bocpd_columns(df):
    series_col = find_column(
        df,
        [
            "coin",
            "symbol",
            "ticker",
            "file",
            "filename",
            "series",
            "name",
        ],
    )

    datetime_col = find_column(
        df,
        [
            "datetime",
            "timestamp",
            "date",
            "time",
            "cp_datetime",
            "Date",
            "Datetime",
            "Timestamp",
        ],
    )

    prob_col = find_column(
        df,
        [
            "bocpd_probability",
            "cp_probability",
            "probability",
            "prob",
            "change_prob",
            "change_probability",
            "p_cp",
            "cp_prob",
        ],
    )

    if prob_col is None:
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        if numeric_cols:
            prob_col = numeric_cols[-1]

    return series_col, datetime_col, prob_col


def infer_pairs_columns(df):
    series_col = find_column(
        df,
        [
            "coin",
            "symbol",
            "ticker",
            "file",
            "filename",
            "series",
            "name",
        ],
    )

    datetime_col = find_column(
        df,
        [
            "cp_datetime",
            "datetime",
            "timestamp",
            "date",
            "time",
            "Date",
            "Datetime",
            "Timestamp",
        ],
    )

    method_col = find_column(
        df,
        [
            "method",
            "algo",
            "algorithm",
            "model",
        ],
    )

    strength_col = find_column(
        df,
        [
            "cp_strength",
            "strength",
            "score",
            "change_strength",
            "cp_score",
        ],
    )

    return series_col, datetime_col, method_col, strength_col


# =============================================================================
# 開始
# =============================================================================

log("=" * 80)
log("BOCPD確率 + pairs変化点 + 元価格時系列 プロット開始")
log("=" * 80)

# =============================================================================
# ファイル読み込み
# =============================================================================

bocpd_csv = load_latest_file(CHANGE_POINT_DIR, BOCPD_PATTERN)
pairs_csv = load_latest_file(CHANGE_POINT_DIR, PAIRS_PATTERN)

log(f"BOCPD CSV: {bocpd_csv}")
log(f"Pairs CSV: {pairs_csv}")

log("BOCPD読み込み中")
bocpd_df = pd.read_csv(bocpd_csv)
log(f"BOCPD行数: {len(bocpd_df):,}")
log(f"BOCPD列数: {len(bocpd_df.columns)}")

log("Pairs読み込み中")
pairs_df = pd.read_csv(pairs_csv)
log(f"Pairs行数: {len(pairs_df):,}")
log(f"Pairs列数: {len(pairs_df.columns)}")

# =============================================================================
# 列推定
# =============================================================================

bocpd_series_col, bocpd_dt_col, bocpd_prob_col = infer_bocpd_columns(bocpd_df)
pairs_series_col, pairs_dt_col, pairs_method_col, pairs_strength_col = infer_pairs_columns(pairs_df)

log("=" * 80)
log("推定列")
log(f"BOCPD series_col : {bocpd_series_col}")
log(f"BOCPD dt_col     : {bocpd_dt_col}")
log(f"BOCPD prob_col   : {bocpd_prob_col}")
log(f"Pairs series_col : {pairs_series_col}")
log(f"Pairs dt_col     : {pairs_dt_col}")
log(f"Pairs method_col : {pairs_method_col}")
log(f"Pairs strength   : {pairs_strength_col}")

if bocpd_series_col is None:
    raise ValueError("BOCPD側の銘柄列が見つかりません。")

if bocpd_dt_col is None:
    raise ValueError("BOCPD側の日時列が見つかりません。")

if bocpd_prob_col is None:
    raise ValueError("BOCPD側の確率列が見つかりません。")

if pairs_series_col is None:
    raise ValueError("pairs側の銘柄列が見つかりません。")

if pairs_dt_col is None:
    raise ValueError("pairs側の日時列が見つかりません。")

# =============================================================================
# 型変換
# =============================================================================

bocpd_df[bocpd_dt_col] = pd.to_datetime(bocpd_df[bocpd_dt_col], errors="coerce")
bocpd_df[bocpd_prob_col] = pd.to_numeric(bocpd_df[bocpd_prob_col], errors="coerce")

pairs_df[pairs_dt_col] = pd.to_datetime(pairs_df[pairs_dt_col], errors="coerce")

if pairs_strength_col is not None:
    pairs_df[pairs_strength_col] = pd.to_numeric(pairs_df[pairs_strength_col], errors="coerce")

bocpd_df = bocpd_df.dropna(subset=[bocpd_dt_col, bocpd_prob_col])
pairs_df = pairs_df.dropna(subset=[pairs_dt_col])

# =============================================================================
# CoinGecko元CSV索引
# =============================================================================

log("=" * 80)
log("CoinGecko元CSV検索")

if not COINGECKO_DIR.exists():
    raise FileNotFoundError(f"元時系列ディレクトリが見つかりません: {COINGECKO_DIR}")

coingecko_csv_files = sorted(COINGECKO_DIR.rglob("*.csv"))
coingecko_file_index = build_file_index(coingecko_csv_files)

log(f"元時系列CSV数: {len(coingecko_csv_files):,}")

# =============================================================================
# 対象銘柄選定
# BOCPD確率の最大値が高い順
# =============================================================================

target_df = (
    bocpd_df
    .groupby(bocpd_series_col)
    .agg(
        bocpd_max_prob=(bocpd_prob_col, "max"),
        bocpd_mean_prob=(bocpd_prob_col, "mean"),
        bocpd_count=(bocpd_prob_col, "size"),
    )
    .reset_index()
    .sort_values("bocpd_max_prob", ascending=False)
    .head(PLOT_TOP_N)
    .reset_index(drop=True)
)

target_df.insert(0, "rank", np.arange(1, len(target_df) + 1))

target_csv = OUTPUT_DIR / f"bocpd_top{PLOT_TOP_N}_targets_{RUN_ID}.csv"
target_df.to_csv(target_csv, index=False, encoding=ENCODING_OUT)
log(f"対象銘柄CSV保存: {target_csv}")

log("=" * 80)
log("プロット対象")
print(target_df.to_string(index=False))

# =============================================================================
# プロット
# =============================================================================

saved = 0
skipped = 0

for _, row in target_df.iterrows():

    rank = int(row["rank"])
    series_name = row[bocpd_series_col]

    log("=" * 80)
    log(f"[{rank}/{PLOT_TOP_N}] プロット開始: {series_name}")

    b = bocpd_df[bocpd_df[bocpd_series_col].astype(str) == str(series_name)].copy()
    p = pairs_df[pairs_df[pairs_series_col].astype(str) == str(series_name)].copy()

    if len(b) == 0:
        log(f"[skip] BOCPDデータなし: {series_name}")
        skipped += 1
        continue

    b = b.sort_values(bocpd_dt_col)
    b_plot = downsample_df(b, MAX_POINTS_PER_SERIES)

    price_df = None
    price_dt_col = None
    price_col = None

    ts_file = find_timeseries_file(series_name, coingecko_csv_files, coingecko_file_index)

    if ts_file is not None:
        try:
            price_df, price_dt_col, price_col = load_price_csv(ts_file)
            price_df = downsample_df(price_df, MAX_POINTS_PER_SERIES)
        except Exception as e:
            log(f"[warn] 価格CSV読み込み失敗: {series_name} / {e}")
            price_df = None
    else:
        log(f"[warn] 元価格CSVが見つかりません: {series_name}")

    fig, axes = plt.subplots(
        nrows=3,
        ncols=1,
        figsize=(16, 10),
        sharex=False,
        gridspec_kw={"height_ratios": [2.0, 1.3, 0.8]},
    )

    ax_price = axes[0]
    ax_prob = axes[1]
    ax_event = axes[2]

    # -------------------------------------------------------------------------
    # 上段: 価格
    # -------------------------------------------------------------------------
    if price_df is not None and len(price_df) > 0:
        y = price_df[price_col].copy()
        ylabel = price_col

        if NORMALIZE_PRICE:
            first_valid = y.dropna().iloc[0]
            if first_valid != 0:
                y = y / first_valid
                ylabel = f"{price_col} normalized"

        ax_price.plot(price_df[price_dt_col], y)
        ax_price.set_ylabel(ylabel)

        if USE_LOG_PRICE:
            ax_price.set_yscale("log")

        ax_price.set_title(f"Rank {rank}: {series_name} - Price / BOCPD Probability / Detected Pairs")
    else:
        ax_price.set_title(f"Rank {rank}: {series_name} - Price not found")
        ax_price.text(
            0.5,
            0.5,
            "Price CSV not found",
            ha="center",
            va="center",
            transform=ax_price.transAxes,
        )

    ax_price.grid(True, alpha=0.3)

    # -------------------------------------------------------------------------
    # 中段: BOCPD確率
    # -------------------------------------------------------------------------
    ax_prob.plot(b_plot[bocpd_dt_col], b_plot[bocpd_prob_col])
    ax_prob.axhline(BOCPD_THRESHOLD, linestyle="--")
    ax_prob.set_ylabel("BOCPD probability")
    ax_prob.set_ylim(
        max(0, min(0, b_plot[bocpd_prob_col].min())),
        min(1.05, max(1.0, b_plot[bocpd_prob_col].max() * 1.05)),
    )
    ax_prob.grid(True, alpha=0.3)

    # -------------------------------------------------------------------------
    # 下段: pairsイベント
    # -------------------------------------------------------------------------
    if len(p) > 0:
        p = p.sort_values(pairs_dt_col)

        if pairs_method_col is not None:
            methods = sorted(p[pairs_method_col].dropna().astype(str).unique())
            method_to_y = {m: i + 1 for i, m in enumerate(methods)}

            for m in methods:
                pm = p[p[pairs_method_col].astype(str) == m]
                y = method_to_y[m]
                ax_event.vlines(pm[pairs_dt_col], y - 0.35, y + 0.35)

            ax_event.set_yticks(list(method_to_y.values()))
            ax_event.set_yticklabels(list(method_to_y.keys()), fontsize=8)
            ax_event.set_ylabel("method")

        else:
            ax_event.vlines(p[pairs_dt_col], 0, 1)
            ax_event.set_yticks([0.5])
            ax_event.set_yticklabels(["pairs"])
            ax_event.set_ylabel("pairs")

        ax_event.set_title(f"Detected change points: {len(p):,}")

    else:
        ax_event.text(
            0.5,
            0.5,
            "No detected pairs for this series",
            ha="center",
            va="center",
            transform=ax_event.transAxes,
        )
        ax_event.set_yticks([])
        ax_event.set_title("Detected change points: 0")

    ax_event.grid(True, alpha=0.3)
    ax_event.set_xlabel("datetime")

    # -------------------------------------------------------------------------
    # pairsの変化点を価格・BOCPDにも薄く重ねる
    # -------------------------------------------------------------------------
    if len(p) > 0:
        cp_times = p[pairs_dt_col].dropna().tolist()

        for t in cp_times:
            ax_price.axvline(t, alpha=0.12)
            ax_prob.axvline(t, alpha=0.12)

    plt.tight_layout()

    out_png = OUTPUT_DIR / f"rank{rank:02d}_{safe_name(series_name)}_bocpd_pairs_price_{RUN_ID}.png"

    if SAVE_PLOTS:
        plt.savefig(out_png, dpi=PLOT_DPI, bbox_inches="tight")
        log(f"保存: {out_png}")

    if SHOW_PLOTS:
        plt.show()

    plt.close()

    saved += 1

# =============================================================================
# 完了
# =============================================================================

log("=" * 80)
log("完了")
log(f"保存数: {saved}")
log(f"スキップ数: {skipped}")
log(f"出力ディレクトリ: {OUTPUT_DIR}")
log(f"ログ: {LOG_FILE}")
log("=" * 80)
```
