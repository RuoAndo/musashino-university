```python
# -*- coding: utf-8 -*-

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# ==========================
# 設定
# ==========================
CSV_PATH = Path(
    r"D:\musashino-university\finance\change_point_output"
    r"\change_point_daily_counts_pivot_25combos_20260510_142302.csv"
)

OUT_PNG = CSV_PATH.parent / "daily_change_point_25combos_overlay.png"

FIGSIZE = (22, 10)

# ==========================
# 読み込み
# ==========================
df = pd.read_csv(CSV_PATH, encoding="utf-8-sig")

df["date"] = pd.to_datetime(df["date"])

df = df.sort_values("date")

# ==========================
# プロット
# ==========================
plt.figure(figsize=FIGSIZE)

combo_cols = [c for c in df.columns if c != "date"]

for col in combo_cols:
    plt.plot(
        df["date"],
        df[col],
        linewidth=1,
        alpha=0.8,
        label=col
    )

plt.title("Daily Change Point Counts by 25 Method/Model Combos")

plt.xlabel("Date")
plt.ylabel("Change Point Count")

plt.grid(True)

plt.legend(
    bbox_to_anchor=(1.02, 1),
    loc="upper left",
    fontsize=8
)

plt.tight_layout()

# 保存
plt.savefig(
    OUT_PNG,
    dpi=150,
    bbox_inches="tight"
)

# 表示
plt.show()

print("保存:", OUT_PNG)
```
