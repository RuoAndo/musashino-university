```python
# -*- coding: utf-8 -*-

from pathlib import Path
from datetime import datetime
import time

import pandas as pd
import matplotlib.pyplot as plt


# ==========================
# 設定
# ==========================
CSV_PATH = Path(
    r"D:\musashino-university\finance\change_point_output"
    r"\change_point_daily_counts_pivot_25combos_20260510_142302.csv"
)

OUTPUT_DIR = CSV_PATH.parent

OVERLAY_PNG = OUTPUT_DIR / "daily_change_point_25combos_overlay.png"

INDIVIDUAL_DIR = OUTPUT_DIR / "plots_25combos_individual"

FIGSIZE_OVERLAY = (22, 10)
FIGSIZE_SINGLE = (16, 6)

DPI = 150

SHOW_OVERLAY = True       # 重ね描画を表示する
SHOW_INDIVIDUAL = True    # 個別25枚を表示する

SAVE_ENCODING = "utf-8-sig"


# ==========================
# ログ
# ==========================
START_TIME = time.time()

def log(msg):
    elapsed = time.time() - START_TIME
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"[{now} | +{elapsed:,.1f}s] {msg}")


# ==========================
# 準備
# ==========================
INDIVIDUAL_DIR.mkdir(parents=True, exist_ok=True)

log("CSV読み込み開始")

df = pd.read_csv(CSV_PATH, encoding=SAVE_ENCODING)

df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

combo_cols = [c for c in df.columns if c != "date"]

log(f"読み込み完了: {CSV_PATH}")
log(f"対象系列数: {len(combo_cols)}")


# ==========================
# 1. 25系列の重ね描画
# ==========================
log("=" * 70)
log("25系列重ね描画を開始")

plt.figure(figsize=FIGSIZE_OVERLAY)

for col in combo_cols:
    plt.plot(
        df["date"],
        df[col],
        linewidth=1,
        alpha=0.8,
        label=col
    )

plt.title("Daily Change Point Counts - 25 Combos Overlay")
plt.xlabel("Date")
plt.ylabel("Change Point Count")
plt.grid(True)

plt.legend(
    bbox_to_anchor=(1.02, 1),
    loc="upper left",
    fontsize=8
)

plt.tight_layout()

plt.savefig(
    OVERLAY_PNG,
    dpi=DPI,
    bbox_inches="tight"
)

log(f"保存完了: {OVERLAY_PNG}")

if SHOW_OVERLAY:
    plt.show()

plt.close()


# ==========================
# 2. 25系列を個別に描画
# ==========================
log("=" * 70)
log("個別プロットを開始")

for i, col in enumerate(combo_cols, start=1):

    log(f"[{i}/{len(combo_cols)}] {col}")

    plt.figure(figsize=FIGSIZE_SINGLE)

    plt.plot(
        df["date"],
        df[col],
        linewidth=2
    )

    plt.title(f"Daily Change Point Counts - {col}")
    plt.xlabel("Date")
    plt.ylabel("Change Point Count")
    plt.grid(True)

    plt.tight_layout()

    out_png = INDIVIDUAL_DIR / f"{col}_daily_change_points.png"

    plt.savefig(
        out_png,
        dpi=DPI
    )

    log(f"保存完了: {out_png}")

    if SHOW_INDIVIDUAL:
        plt.show()

    plt.close()


# ==========================
# 3. 最大値ランキング
# ==========================
log("=" * 70)
log("最大値ランキングを作成")

ranking = []

for col in combo_cols:
    max_idx = df[col].idxmax()
    ranking.append({
        "combo": col,
        "max_count": int(df.loc[max_idx, col]),
        "max_date": df.loc[max_idx, "date"].strftime("%Y-%m-%d"),
        "total_count": int(df[col].sum())
    })

ranking_df = pd.DataFrame(ranking)

ranking_df = ranking_df.sort_values(
    ["max_count", "total_count"],
    ascending=False
)

RANKING_CSV = OUTPUT_DIR / "daily_change_point_25combos_ranking.csv"

ranking_df.to_csv(
    RANKING_CSV,
    index=False,
    encoding=SAVE_ENCODING
)

log(f"ランキングCSV保存完了: {RANKING_CSV}")

print()
print("=== 最大値ランキング ===")
print(ranking_df.to_string(index=False))


# ==========================
# 終了
# ==========================
log("=" * 70)
log("処理完了")
log(f"重ね描画PNG: {OVERLAY_PNG}")
log(f"個別PNG保存先: {INDIVIDUAL_DIR}")
log(f"ランキングCSV: {RANKING_CSV}")
log(f"総処理時間: {time.time() - START_TIME:,.1f}s")
```
