```python
import pandas as pd

file_path = r"d:\musashino-university\finance\change_point_pairs.csv"

df = pd.read_csv(
    file_path,
    skiprows=1,
    names=["coin", "datetime", "score"]
)

df["datetime"] = pd.to_datetime(
    df["datetime"],
    format="%Y-%m-%d %H:%M:%S",
    errors="coerce"
)

df = df.dropna(subset=["datetime"])

df["date"] = df["datetime"].dt.date

daily_counts = (
    df.groupby("date")
      .size()
      .reset_index(name="count")
)

print(daily_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"d:\musashino-university\finance\change_point_pairs.csv"

df = pd.read_csv(
    file_path,
    skiprows=1,
    names=["coin", "datetime", "score"]
)

df["datetime"] = pd.to_datetime(
    df["datetime"],
    format="%Y-%m-%d %H:%M:%S",
    errors="coerce"
)

df = df.dropna(subset=["datetime"])

# 日付だけ取り出し
df["date"] = df["datetime"].dt.date

# 日ごとの件数
daily_counts = df.groupby("date").size()

# プロット
plt.figure()
daily_counts.plot(kind="bar")

plt.xlabel("date")
plt.ylabel("count")
plt.title("Daily Change Points")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
