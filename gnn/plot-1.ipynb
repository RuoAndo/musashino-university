```python
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

png_files = sorted(
    Path(".").glob("edge_count_variable_gcn*.png")
)

cols = 3
rows = (len(png_files) + cols - 1) // cols

fig, axes = plt.subplots(
    rows,
    cols,
    figsize=(20, rows * 4),
)

axes = axes.flatten()

for ax, f in zip(axes, png_files):

    img = mpimg.imread(f)

    ax.imshow(img)

    ax.set_title(f.name, fontsize=8)

    ax.axis("off")

for ax in axes[len(png_files):]:
    ax.axis("off")

plt.tight_layout()

plt.show()
```
