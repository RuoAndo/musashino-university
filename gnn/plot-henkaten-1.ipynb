```python
import math
from pathlib import Path

import matplotlib.pyplot as plt

# =====================================
# 設定
# =====================================
BASE_DIR = Path(".")
FIGSIZE_PER_IMAGE = 4

# =====================================
# PNG取得
# =====================================
png_files = sorted(BASE_DIR.glob("*.png"))

if len(png_files) == 0:
    raise FileNotFoundError("PNGファイルがありません。")

print(f"PNG数: {len(png_files)}")

# =====================================
# まず3列一覧表示
# =====================================
cols = 3
rows = math.ceil(len(png_files) / cols)

fig, axes = plt.subplots(
    rows,
    cols,
    figsize=(cols * FIGSIZE_PER_IMAGE,
             rows * FIGSIZE_PER_IMAGE)
)

# axesを1次元化
if len(png_files) == 1:
    axes = [axes]
else:
    axes = axes.ravel()

for ax, path in zip(axes, png_files):

    try:
        img = plt.imread(path)

        ax.imshow(img)
        ax.set_title(path.name, fontsize=8)
        ax.axis("off")

        print(f"一覧表示: {path.name}")

    except Exception as e:

        ax.set_title(f"ERROR\n{path.name}")
        ax.axis("off")

        print(f"失敗: {path.name} | {e}")

# 余り非表示
for ax in axes[len(png_files):]:
    ax.axis("off")

plt.tight_layout()
plt.show()

# =====================================
# その後、1枚ずつ大きく表示
# =====================================
for i, path in enumerate(png_files, start=1):

    try:
        img = plt.imread(path)

        plt.figure(figsize=(12, 8))

        plt.imshow(img)

        plt.title(
            f"[{i}/{len(png_files)}] {path.name}",
            fontsize=12
        )

        plt.axis("off")

        print(f"単独表示: {path.name}")

        plt.tight_layout()
        plt.show()

    except Exception as e:

        print(f"単独表示失敗: {path.name} | {e}")
```
