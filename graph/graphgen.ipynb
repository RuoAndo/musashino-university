```python
# -*- coding: utf-8 -*-

"""
6種類のトポロジー例を生成・可視化する完全版

対象
1. ヘテログラフ
2. 二部グラフ
3. 多次元グラフ
4. 符号付きグラフ
5. ハイパーグラフ
6. ダイナミックグラフ

特徴
・日本語文字化け対策
・PNG保存対応
・表示/非表示を切り替え可能
・ダイナミックグラフは時系列変化も別図保存
"""

from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import networkx as nx


# =========================================================
# 設定
# =========================================================
OUTPUT_DIR = Path("./graph_topology_output")
SHOW_FIGURES = True
SAVE_FIGURES = True
DPI = 160


# =========================================================
# 日本語フォント設定
# =========================================================
def setup_japanese_font():
    candidates = [
        "IPAexGothic",
        "IPAGothic",
        "Noto Sans CJK JP",
        "Noto Sans JP",
        "Yu Gothic",
        "Meiryo",
        "MS Gothic",
        "Hiragino Sans",
        "TakaoGothic",
    ]

    available_fonts = {f.name for f in fm.fontManager.ttflist}

    for name in candidates:
        if name in available_fonts:
            plt.rcParams["font.family"] = name
            plt.rcParams["axes.unicode_minus"] = False
            print(f"[INFO] 日本語フォントを設定しました: {name}")
            return name

    plt.rcParams["axes.unicode_minus"] = False
    print("[WARN] 日本語フォントが見つかりませんでした。環境によっては文字化けする可能性があります。")
    return None


# =========================================================
# 1. ヘテログラフ
# =========================================================
def build_heterograph():
    G = nx.DiGraph()

    # ノード種別つき
    G.add_node("著者A", ntype="author")
    G.add_node("著者B", ntype="author")
    G.add_node("論文1", ntype="paper")
    G.add_node("論文2", ntype="paper")
    G.add_node("キーワードX", ntype="keyword")
    G.add_node("キーワードY", ntype="keyword")

    # 関係種別つき
    G.add_edge("著者A", "論文1", etype="執筆")
    G.add_edge("著者B", "論文2", etype="執筆")
    G.add_edge("論文1", "キーワードX", etype="含む")
    G.add_edge("論文1", "キーワードY", etype="含む")
    G.add_edge("論文2", "キーワードY", etype="含む")
    G.add_edge("著者A", "著者B", etype="共著")

    return G


def draw_heterograph(ax):
    G = build_heterograph()
    pos = nx.spring_layout(G, seed=42, k=1.2)

    color_map = {
        "author": "skyblue",
        "paper": "lightgreen",
        "keyword": "salmon",
    }

    node_colors = [color_map[G.nodes[n]["ntype"]] for n in G.nodes()]

    nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=1600, ax=ax)
    nx.draw_networkx_labels(G, pos, font_size=10, ax=ax)
    nx.draw_networkx_edges(G, pos, arrows=True, arrowstyle="-|>", arrowsize=18, ax=ax)

    edge_labels = {(u, v): d["etype"] for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, ax=ax)

    ax.set_title("1. ヘテログラフ")
    ax.axis("off")


# =========================================================
# 2. 二部グラフ
# =========================================================
def build_bipartite_graph():
    G = nx.Graph()

    left_nodes = ["利用者1", "利用者2", "利用者3"]
    right_nodes = ["商品A", "商品B", "商品C", "商品D"]

    G.add_nodes_from(left_nodes, bipartite=0)
    G.add_nodes_from(right_nodes, bipartite=1)

    edges = [
        ("利用者1", "商品A"),
        ("利用者1", "商品B"),
        ("利用者2", "商品B"),
        ("利用者2", "商品C"),
        ("利用者3", "商品C"),
        ("利用者3", "商品D"),
    ]
    G.add_edges_from(edges)

    return G, left_nodes, right_nodes


def draw_bipartite_graph(ax):
    G, left_nodes, right_nodes = build_bipartite_graph()

    pos = {}
    for i, n in enumerate(left_nodes):
        pos[n] = (0, -i)
    for i, n in enumerate(right_nodes):
        pos[n] = (2.5, -i)

    node_colors = ["skyblue" if n in left_nodes else "lightgreen" for n in G.nodes()]

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color=node_colors,
        node_size=1600,
        font_size=10,
        ax=ax,
    )

    ax.set_title("2. 二部グラフ")
    ax.axis("off")


# =========================================================
# 3. 多次元グラフ
# =========================================================
def build_multilayer_graph():
    friendship = nx.Graph()
    message = nx.Graph()

    nodes = ["A", "B", "C", "D"]
    friendship.add_nodes_from(nodes)
    message.add_nodes_from(nodes)

    friendship.add_edges_from([("A", "B"), ("B", "C"), ("C", "D")])
    message.add_edges_from([("A", "C"), ("A", "D"), ("B", "D")])

    return friendship, message


def draw_multilayer_graph(ax):
    G1, G2 = build_multilayer_graph()
    pos = nx.circular_layout(G1)

    nx.draw_networkx_nodes(G1, pos, node_color="lightgray", node_size=1600, ax=ax)
    nx.draw_networkx_labels(G1, pos, font_size=10, ax=ax)

    nx.draw_networkx_edges(G1, pos, edge_color="blue", width=2.5, ax=ax)
    nx.draw_networkx_edges(G2, pos, edge_color="red", width=2.5, style="dashed", ax=ax)

    ax.set_title("3. 多次元グラフ\n青実線=友情層 / 赤破線=メッセージ層")
    ax.axis("off")


# =========================================================
# 4. 符号付きグラフ
# =========================================================
def build_signed_graph():
    G = nx.Graph()

    G.add_nodes_from(["A", "B", "C", "D", "E"])
    signed_edges = [
        ("A", "B", +1),
        ("A", "C", -1),
        ("B", "C", +1),
        ("B", "D", -1),
        ("C", "D", +1),
        ("D", "E", -1),
    ]

    for u, v, sign in signed_edges:
        G.add_edge(u, v, sign=sign)

    return G


def draw_signed_graph(ax):
    G = build_signed_graph()
    pos = nx.spring_layout(G, seed=7, k=1.0)

    pos_edges = [(u, v) for u, v, d in G.edges(data=True) if d["sign"] > 0]
    neg_edges = [(u, v) for u, v, d in G.edges(data=True) if d["sign"] < 0]

    nx.draw_networkx_nodes(G, pos, node_color="lightyellow", node_size=1600, ax=ax)
    nx.draw_networkx_labels(G, pos, font_size=10, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=pos_edges, edge_color="blue", width=2.5, ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=neg_edges, edge_color="red", width=2.5, style="dashed", ax=ax)

    edge_labels = {(u, v): "+" if d["sign"] > 0 else "-" for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, ax=ax)

    ax.set_title("4. 符号付きグラフ\n青=正エッジ / 赤破線=負エッジ")
    ax.axis("off")


# =========================================================
# 5. ハイパーグラフ
# =========================================================
def build_hypergraph_as_bipartite():
    """
    ハイパーエッジを補助ノードで表す。
    例:
      頂点 v1,v2,v3...
      ハイパーエッジ e1,e2...
    """
    G = nx.Graph()

    vertices = ["v1", "v2", "v3", "v4", "v5"]
    hyperedges = {
        "e1": ["v1", "v2", "v3"],
        "e2": ["v2", "v4"],
        "e3": ["v3", "v4", "v5"],
    }

    for v in vertices:
        G.add_node(v, kind="vertex")
    for e in hyperedges:
        G.add_node(e, kind="hyperedge")

    for e, vs in hyperedges.items():
        for v in vs:
            G.add_edge(e, v)

    return G, vertices, list(hyperedges.keys())


def draw_hypergraph(ax):
    G, vertices, hyperedge_nodes = build_hypergraph_as_bipartite()

    pos = {}
    for i, v in enumerate(vertices):
        pos[v] = (0, -i)
    for i, e in enumerate(hyperedge_nodes):
        pos[e] = (2.5, -i - 0.5)

    node_colors = ["lightgreen" if n in vertices else "orange" for n in G.nodes()]

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color=node_colors,
        node_size=1600,
        font_size=10,
        ax=ax,
    )

    ax.set_title("5. ハイパーグラフ\n橙ノード=ハイパーエッジ")
    ax.axis("off")


# =========================================================
# 6. ダイナミックグラフ
# =========================================================
def build_dynamic_graph_sequence():
    g1 = nx.Graph()
    g2 = nx.Graph()
    g3 = nx.Graph()

    nodes = ["A", "B", "C", "D"]
    for g in (g1, g2, g3):
        g.add_nodes_from(nodes)

    g1.add_edges_from([("A", "B"), ("B", "C")])
    g2.add_edges_from([("A", "B"), ("B", "C"), ("C", "D")])
    g3.add_edges_from([("A", "D"), ("B", "C"), ("C", "D")])

    return [g1, g2, g3]


def draw_dynamic_graph(ax):
    graphs = build_dynamic_graph_sequence()
    G = graphs[-1]
    pos = nx.circular_layout(G)

    nx.draw_networkx_nodes(G, pos, node_color="violet", node_size=1600, ax=ax)
    nx.draw_networkx_labels(G, pos, font_size=10, ax=ax)
    nx.draw_networkx_edges(G, pos, width=2.5, ax=ax)

    ax.set_title("6. ダイナミックグラフ\n最終時刻 t=3")
    ax.axis("off")


def plot_dynamic_sequence(output_dir: Path):
    graphs = build_dynamic_graph_sequence()
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    pos = nx.circular_layout(graphs[0])

    for i, G in enumerate(graphs):
        nx.draw_networkx_nodes(G, pos, node_color="violet", node_size=1300, ax=axes[i])
        nx.draw_networkx_labels(G, pos, font_size=10, ax=axes[i])
        nx.draw_networkx_edges(G, pos, width=2.5, ax=axes[i])
        axes[i].set_title(f"時刻 t={i+1}")
        axes[i].axis("off")

    plt.tight_layout()

    if SAVE_FIGURES:
        out_path = output_dir / "06_dynamic_graph_sequence.png"
        fig.savefig(out_path, dpi=DPI, bbox_inches="tight")
        print(f"[INFO] 保存しました: {out_path}")

    if SHOW_FIGURES:
        plt.show()
    else:
        plt.close(fig)


# =========================================================
# メイン図
# =========================================================
def plot_all_graphs(output_dir: Path):
    fig, axes = plt.subplots(2, 3, figsize=(17, 10))

    draw_heterograph(axes[0, 0])
    draw_bipartite_graph(axes[0, 1])
    draw_multilayer_graph(axes[0, 2])
    draw_signed_graph(axes[1, 0])
    draw_hypergraph(axes[1, 1])
    draw_dynamic_graph(axes[1, 2])

    plt.tight_layout()

    if SAVE_FIGURES:
        out_path = output_dir / "00_all_topologies.png"
        fig.savefig(out_path, dpi=DPI, bbox_inches="tight")
        print(f"[INFO] 保存しました: {out_path}")

    if SHOW_FIGURES:
        plt.show()
    else:
        plt.close(fig)


# =========================================================
# 個別保存したい場合の補助
# =========================================================
def save_individual_graphs(output_dir: Path):
    drawers = [
        ("01_heterograph.png", draw_heterograph),
        ("02_bipartite_graph.png", draw_bipartite_graph),
        ("03_multilayer_graph.png", draw_multilayer_graph),
        ("04_signed_graph.png", draw_signed_graph),
        ("05_hypergraph.png", draw_hypergraph),
        ("06_dynamic_graph_final.png", draw_dynamic_graph),
    ]

    for filename, drawer in drawers:
        fig, ax = plt.subplots(figsize=(6, 5))
        drawer(ax)
        plt.tight_layout()

        if SAVE_FIGURES:
            out_path = output_dir / filename
            fig.savefig(out_path, dpi=DPI, bbox_inches="tight")
            print(f"[INFO] 保存しました: {out_path}")

        if SHOW_FIGURES:
            plt.show()
        else:
            plt.close(fig)


# =========================================================
# 実行
# =========================================================
def main():
    OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

    setup_japanese_font()

    plot_all_graphs(OUTPUT_DIR)
    save_individual_graphs(OUTPUT_DIR)
    plot_dynamic_sequence(OUTPUT_DIR)

    print("[INFO] 完了しました。")


if __name__ == "__main__":
    main()
```
