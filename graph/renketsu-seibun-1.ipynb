```python
import networkx as nx
import matplotlib.pyplot as plt

# =========================================
# 1. 連結成分が1のグラフ
# =========================================
G1 = nx.Graph()
G1.add_edges_from([
    (1, 2), (2, 3), (3, 4), (4, 5), (2, 5)
])

# =========================================
# 2. 連結成分が2のグラフ
# =========================================
G2 = nx.Graph()
G2.add_edges_from([
    (1, 2), (2, 3),      # 第1成分
    (5, 6), (6, 7), (7, 5)  # 第2成分
])

# =========================================
# 描画
# =========================================
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# 左: 連結成分が1
pos1 = nx.spring_layout(G1, seed=42)
nx.draw(
    G1, pos1, ax=axes[0],
    with_labels=True,
    node_size=800,
    font_size=12
)
axes[0].set_title("連結成分が1のグラフ")
axes[0].axis("off")

# 右: 連結成分が2
pos2 = {
    1: (-1.5, 0), 2: (-1, 0.5), 3: (-0.5, 0),
    5: (0.8, 0), 6: (1.3, 0.5), 7: (1.3, -0.5)
}
nx.draw(
    G2, pos2, ax=axes[1],
    with_labels=True,
    node_size=800,
    font_size=12
)
axes[1].set_title("連結成分が2のグラフ")
axes[1].axis("off")

plt.tight_layout()
plt.show()

# =========================================
# 確認
# =========================================
print("G1 の連結成分数:", nx.number_connected_components(G1))
print("G2 の連結成分数:", nx.number_connected_components(G2))
```
