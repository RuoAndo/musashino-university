```python
import pymc as pm
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
data = iris.data
species = iris.target

# Versicolor1
vers = data[species == 1, 2]  # 

with pm.Model() as model:
    mu = pm.Normal("mu", mu=5, sigma=2)
    sigma = pm.HalfNormal("sigma", sigma=1)
    y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=vers)

    trace = pm.sample(2000, tune=1000)

pm.summary(trace)
```
