```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import os, zipfile, re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# =====  =====
ZIP_PATH = "./isolation_forest.zip"     # 
EXTRACT_DIR = "./isolation_forest_all"  # 

# ===== ZIP =====
os.makedirs(EXTRACT_DIR, exist_ok=True)
with zipfile.ZipFile(ZIP_PATH, "r") as z:
    z.extractall(EXTRACT_DIR)

# ===== CSV =====
csv_files = []
for root, _, files in os.walk(EXTRACT_DIR):
    for f in files:
        if f.lower().endswith(".csv"):
            csv_files.append(os.path.join(root, f))
csv_files.sort()
print(f"CSV: {len(csv_files)}")

# =====  =====
def detect_lat_lon_columns(columns):
    # 
    lat = next((c for c in columns if c.lower() in ["latitude", "lat", "y", "lat_deg"]), None)
    lon = next((c for c in columns if c.lower() in ["longitude", "lon", "lng", "x", "long", "lon_deg"]), None)
    # 
    if lat is None:
        lat = next((c for c in columns if "lat" in c.lower()), None)
    if lon is None:
        lon = next((c for c in columns if ("lon" in c.lower()) or ("lng" in c.lower()) or ("longi" in c.lower())), None)
    return lat, lon

# ===== IPv4  =====
IPV4_RE = re.compile(r"\b(?:(?:25[0-5]|2[0-4]\d|1?\d?\d)\.){3}(?:25[0-5]|2[0-4]\d|1?\d?\d)\b")

def find_ipv4_in_text(text):
    if not isinstance(text, str):
        return None
    m = IPV4_RE.search(text)
    return m.group(0) if m else None

def detect_ip_column(df: pd.DataFrame):
    """IP None"""
    cols = list(df.columns)
    lower = [c.lower() for c in cols]

    # 1) 
    name_prefs = ["ip", "ip_address", "source_ip", "src_ip", "dst_ip", "ipaddr", "address", "remote_ip", "client_ip"]
    for p in name_prefs:
        if p in lower:
            return cols[lower.index(p)]

    # 2) IPv4
    text_cols = [c for c in cols if df[c].dtype == "object"]
    if text_cols:
        sample_n = min(200, len(df))
        best_col, best_rate = None, 0.0
        for c in text_cols:
            series = df[c].astype(str).head(sample_n)
            hits = series.apply(lambda s: IPV4_RE.search(s) is not None).mean()
            if hits > best_rate:
                best_rate, best_col = hits, c
        # : 10%
        if best_col is not None and best_rate >= 0.10:
            return best_col

    # 3)  None
    return None

# =====  =====
for path in csv_files:
    try:
        df = pd.read_csv(path)
    except Exception as e:
        print(f"[SKIP]  {path}: {e}")
        continue

    lat_col, lon_col = detect_lat_lon_columns(df.columns)
    if not lat_col or not lon_col:
        print(f"[SKIP] {os.path.basename(path)}: / -> {list(df.columns)}")
        continue

    # ---- ----
    df = df.copy()
    df[lat_col] = pd.to_numeric(df[lat_col], errors="coerce")
    df[lon_col] = pd.to_numeric(df[lon_col], errors="coerce")
    df = df.dropna(subset=[lat_col, lon_col])
    df = df[(df[lat_col].between(-90, 90)) & (df[lon_col].between(-180, 180))]
    if len(df) < 5:
        print(f"[SKIP] {os.path.basename(path)}:  (n={len(df)})")
        continue

    # ---- LOF1%----
    X = df[[lon_col, lat_col]].to_numpy()
    n = len(X)
    n_neighbors = max(10, min(35, int(n * 0.01)))
    if n_neighbors >= n:
        n_neighbors = max(5, n - 1)

    lof = LocalOutlierFactor(n_neighbors=n_neighbors, contamination=0.01)
    y_pred = lof.fit_predict(X)              # 1=inlier, -1=outlier
    scores = lof.negative_outlier_factor_    # 

    is_out = (y_pred == -1)
    inliers = ~is_out

    # ---- IP   ----
    ip_col = detect_ip_column(df)
    ip_series = None
    if ip_col:
        ip_series = df[ip_col].astype(str)
    else:
        # IPv4
        text_cols = [c for c in df.columns if df[c].dtype == "object"]
        if text_cols:
            concat_text = df[text_cols].astype(str).agg(" ".join, axis=1)
            ip_series = concat_text.apply(lambda s: find_ipv4_in_text(s) or "")

    # ---- outlier IP /----
    out_ips = []
    if ip_series is not None:
        for v in ip_series[is_out].tolist():
            ip = find_ipv4_in_text(v) if isinstance(v, str) else None
            if ip is None and isinstance(v, str) and IPV4_RE.fullmatch(v):
                ip = v
            if ip and ip not in out_ips:
                out_ips.append(ip)

    fname = os.path.basename(path).replace(".csv", "")
    print(f"\n {fname}")
    print(f"  : {n} / : {int(inliers.sum())} / : {int(is_out.sum())} ( {is_out.sum()/n*100:.1f}% )")
    print(f"  n_neighbors={n_neighbors}  score(mean): inlier={scores[inliers].mean():.4f} / outlier={scores[is_out].mean():.4f}")

    if out_ips:
        print("  outlier IPs (50):")
        for ip in out_ips[:50]:
            print(f"    - {ip}")
        if len(out_ips) > 50:
            print(f"    ... and {len(out_ips)-50} more")
    else:
        print("  (outlier IP)")

    # ----  ----
    plt.figure(figsize=(10, 5))
    plt.scatter(X[inliers, 0], X[inliers, 1], s=8, alpha=0.6, label="Inlier")
    plt.scatter(X[is_out, 0],  X[is_out, 1],  s=40, alpha=0.9, marker="x", label="Outlier")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.title(f"LOF (1% Outliers) - {fname}")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

#####

# =====  =====
# UTF-8
import sys, re
try:
    sys.stdout.reconfigure(encoding="utf-8")
except Exception:
    pass  # 

# --- LOF(1%)CSVIP ---
IPV4_RE = re.compile(r"\b(?:(?:25[0-5]|2[0-4]\d|1?\d?\d)\.){3}(?:25[0-5]|2[0-4]\d|1?\d?\d)\b")

def extract_outlier_ips_for_file(df, is_out):
    """ detect_ip_column / find_ipv4_in_text IP"""
    ip_col = detect_ip_column(df)
    ips = []

    if ip_col:
        # 
        series = df.loc[is_out, ip_col].astype(str)
        for s in series:
            m = IPV4_RE.search(s)
            if m:
                ips.append(m.group(0))
    else:
        # IPv4
        text_cols = [c for c in df.columns if df[c].dtype == "object"]
        if text_cols:
            concat_text = df[text_cols].astype(str).agg(" ".join, axis=1)
            for s in concat_text[is_out]:
                m = IPV4_RE.search(s)
                if m:
                    ips.append(m.group(0))
    return set(ips)

summary = []

for path in csv_files:
    try:
        df = pd.read_csv(path)
    except Exception:
        continue

    lat_col, lon_col = detect_lat_lon_columns(df.columns)
    if not lat_col or not lon_col:
        continue

    # 
    df = df.copy()
    df[lat_col] = pd.to_numeric(df[lat_col], errors="coerce")
    df[lon_col] = pd.to_numeric(df[lon_col], errors="coerce")
    df = df.dropna(subset=[lat_col, lon_col])
    df = df[(df[lat_col].between(-90, 90)) & (df[lon_col].between(-180, 180))]
    if len(df) < 5:
        continue

    X = df[[lon_col, lat_col]].to_numpy()
    n = len(X)
    n_neighbors = max(10, min(35, int(n * 0.01)))
    if n_neighbors >= n:
        n_neighbors = max(5, n - 1)

    # LOF(1%) 
    lof = LocalOutlierFactor(n_neighbors=n_neighbors, contamination=0.01)
    y_pred = lof.fit_predict(X)
    is_out = (y_pred == -1)

    # IP
    outlier_ips = extract_outlier_ips_for_file(df, is_out)
    uniq_count = len(outlier_ips)

    # (YYYYMMDD)
    fname = os.path.basename(path).replace(".csv", "")
    m = re.search(r"(\d{8})", fname)
    date = pd.to_datetime(m.group(1), format="%Y%m%d", errors="coerce") if m else pd.NaT

    summary.append({"file": fname, "date": date, "outlier_ip_count": uniq_count})

# DataFrame
summary_df = pd.DataFrame(summary).dropna(subset=["date"]).sort_values("date")
if not summary_df.empty:
    plt.figure(figsize=(10, 5))
    plt.plot(summary_df["date"], summary_df["outlier_ip_count"], marker="o")
    plt.title("IPLOF 1%")
    plt.xlabel("")
    plt.ylabel("IP")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 
    print(summary_df.to_string(index=False))
else:
    print("IP")
# =====  =====


```
