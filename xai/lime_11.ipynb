```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.vector_ar.var_model import VAR

from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


# =========================================
# 0. 共通関数
# =========================================
def evaluate(y_true, y_pred, name="model"):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"\n[{name}]")
    print("RMSE:", rmse)
    print("MAE :", mae)
    print("R2  :", r2)

    return {"RMSE": rmse, "MAE": mae, "R2": r2}


def plot_line(dates, values, title, ylabel="Value", label=None):
    plt.figure(figsize=(12, 4))
    plt.plot(dates, values, label=label)
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel(ylabel)
    if label is not None:
        plt.legend()
    plt.grid(True)
    plt.show()


def plot_compare(dates, actual, pred, title, ylabel="Value",
                 actual_label="Actual", pred_label="Predicted"):
    plt.figure(figsize=(12, 4))
    plt.plot(dates, actual, label=actual_label)
    plt.plot(dates, pred, label=pred_label)
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()


def plot_scatter_actual_pred(y_true, y_pred, title):
    plt.figure(figsize=(6, 6))
    plt.scatter(y_true, y_pred, alpha=0.7)
    mn = min(np.min(y_true), np.min(y_pred))
    mx = max(np.max(y_true), np.max(y_pred))
    plt.plot([mn, mx], [mn, mx], linestyle="--")
    plt.title(title)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.grid(True)
    plt.show()


def plot_residuals(dates, residuals, title):
    plt.figure(figsize=(12, 4))
    plt.plot(dates, residuals)
    plt.axhline(0, linestyle="--")
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Residual")
    plt.grid(True)
    plt.show()


def plot_hist(data, title, xlabel="Value", bins=20):
    plt.figure(figsize=(8, 4))
    plt.hist(data, bins=bins)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()


def rolling_forecast_ar(train_series, test_series, lags):
    history = list(np.asarray(train_series))
    preds = []
    test_arr = np.asarray(test_series)

    for t in range(len(test_arr)):
        model = AutoReg(history, lags=lags, old_names=False)
        result = model.fit()
        pred = result.predict(start=len(history), end=len(history))[0]
        preds.append(pred)
        history.append(test_arr[t])

    return np.asarray(preds)


def rolling_forecast_arima(train_series, test_series, order=(0, 0, 1)):
    history = list(np.asarray(train_series))
    preds = []
    test_arr = np.asarray(test_series)

    for t in range(len(test_arr)):
        model = ARIMA(history, order=order)
        result = model.fit()
        pred = result.forecast(steps=1)[0]
        preds.append(pred)
        history.append(test_arr[t])

    return np.asarray(preds)


def rolling_forecast_sarima(train_series, test_series,
                            order=(1, 1, 1),
                            seasonal_order=(0, 0, 0, 0)):
    history = list(np.asarray(train_series))
    preds = []
    test_arr = np.asarray(test_series)

    for t in range(len(test_arr)):
        model = SARIMAX(
            history,
            order=order,
            seasonal_order=seasonal_order,
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        result = model.fit(disp=False)
        pred = result.forecast(steps=1)[0]
        preds.append(pred)
        history.append(test_arr[t])

    return np.asarray(preds)


def fit_var_safely(train_df, maxlags=2):
    """
    VAR を安全に fit する。
    maxlags から順に下げて試し、成功したら result を返す。
    """
    maxlags = int(maxlags)
    maxlags = min(maxlags, len(train_df) // 3 if len(train_df) >= 6 else 1)
    maxlags = max(1, maxlags)

    last_error = None
    for lag in range(maxlags, 0, -1):
        try:
            model = VAR(train_df)
            result = model.fit(lag)
            return result, lag
        except Exception as e:
            last_error = e

    raise last_error


def rolling_forecast_var(train_df, test_df, maxlags=2):
    """
    VAR の rolling forecast。
    各時点で安全に fit し、1ステップ先予測する。
    """
    history = train_df.copy()
    preds = []

    for i in range(len(test_df)):
        result, used_lag = fit_var_safely(history, maxlags=maxlags)
        lag_order = result.k_ar

        input_data = history.values[-lag_order:]
        pred = result.forecast(input_data, steps=1)[0]
        preds.append(pred)

        next_actual = test_df.iloc[[i]]
        history = pd.concat([history, next_actual], axis=0)

    pred_df = pd.DataFrame(preds, columns=test_df.columns, index=test_df.index)
    return pred_df


# =========================================
# 1. CSV読み込み
# =========================================
df = pd.read_csv("monthly.csv")

df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

df = df.dropna(subset=["Date", "Price"]).copy()
df = df.sort_values("Date").reset_index(drop=True)

print("元データ件数:", len(df))
print(df.head())


# =========================================
# 2. 期間抽出
# =========================================
df = df[(df["Date"] >= "1950-01-01") & (df["Date"] <= "2020-07-01")].copy()
df = df.reset_index(drop=True)

print("期間抽出後件数:", len(df))


# =========================================
# 3. 元系列確認
# =========================================
plot_line(
    df["Date"],
    df["Price"],
    title="Original Series",
    ylabel="Price"
)


# =========================================
# 4. 季節差分
# =========================================
df["diff"] = df["Price"].diff(12)
df = df.dropna().reset_index(drop=True)

plot_line(
    df["Date"],
    df["diff"],
    title="Seasonal Differenced Series (lag=12)",
    ylabel="Differenced Price"
)


# =========================================
# 5. ACF / PACF
# =========================================
plot_acf(df["diff"], lags=40)
plt.title("ACF of Seasonal Differenced Series")
plt.show()

plot_pacf(df["diff"], lags=40, method="ywm")
plt.title("PACF of Seasonal Differenced Series")
plt.show()


# =========================================
# 6. 特徴量生成
# =========================================
lag_list = [1, 2, 3, 6, 12]
for lag in lag_list:
    df[f"lag_{lag}"] = df["diff"].shift(lag)

df["month"] = df["Date"].dt.month
df["year"] = df["Date"].dt.year

# VAR 用の安全な季節説明変数
df["month_sin"] = np.sin(2 * np.pi * df["month"] / 12)
df["month_cos"] = np.cos(2 * np.pi * df["month"] / 12)

df = df.dropna().reset_index(drop=True)


# =========================================
# 7. X / y
# =========================================
feature_cols = ["lag_1", "lag_2", "lag_3", "lag_6", "lag_12", "month", "year"]

X = df[feature_cols].copy()
y = df["diff"].copy()
dates = df["Date"].copy()


# =========================================
# 8. 時系列分割
# =========================================
split = int(len(df) * 0.8)

X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]
date_train, date_test = dates.iloc[:split], dates.iloc[split:]

plt.figure(figsize=(12, 4))
plt.plot(date_train, y_train, label="Train")
plt.plot(date_test, y_test, label="Test")
plt.axvline(dates.iloc[split], linestyle="--")
plt.title("Train / Test Split on Differenced Series")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 9. Random Forest 学習
# =========================================
rf_model = RandomForestRegressor(
    n_estimators=500,
    random_state=1,
    n_jobs=-1
)
rf_model.fit(X_train, y_train)


# =========================================
# 10. Random Forest 予測
# =========================================
y_train_pred_rf = rf_model.predict(X_train)
y_test_pred_rf = rf_model.predict(X_test)


# =========================================
# 11. Random Forest 評価
# =========================================
rf_train_metrics = evaluate(y_train, y_train_pred_rf, "Random Forest Train")
rf_test_metrics = evaluate(y_test, y_test_pred_rf, "Random Forest Test")


# =========================================
# 12. Random Forest の図
# =========================================
plot_compare(
    date_train, y_train.values, y_train_pred_rf,
    title="Random Forest - Train Fit",
    ylabel="Differenced Price",
    actual_label="Train Actual",
    pred_label="Train Predicted"
)

plot_compare(
    date_test, y_test.values, y_test_pred_rf,
    title="Random Forest - Test Prediction",
    ylabel="Differenced Price",
    actual_label="Test Actual",
    pred_label="Test Predicted"
)

plot_scatter_actual_pred(
    y_test.values,
    y_test_pred_rf,
    title="Random Forest - Actual vs Predicted"
)

rf_residuals = y_test.values - y_test_pred_rf

plot_residuals(
    date_test,
    rf_residuals,
    title="Random Forest - Residuals Over Time"
)

plot_hist(
    rf_residuals,
    title="Random Forest - Residual Histogram",
    xlabel="Residual"
)

plot_line(
    date_test,
    np.abs(rf_residuals),
    title="Random Forest - Absolute Error Over Time",
    ylabel="Absolute Error"
)


# =========================================
# 13. 特徴量重要度
# =========================================
importance = pd.Series(rf_model.feature_importances_, index=feature_cols)
importance = importance.sort_values(ascending=False)

print("\n特徴量重要度")
print(importance)

plt.figure(figsize=(10, 4))
importance.plot(kind="bar")
plt.title("Random Forest - Feature Importance")
plt.ylabel("Importance")
plt.grid(True, axis="y")
plt.show()


# =========================================
# 14. RFE
# =========================================
rfe_base_model = RandomForestRegressor(
    n_estimators=300,
    random_state=1,
    n_jobs=-1
)

rfe = RFE(rfe_base_model, n_features_to_select=4)
rfe.fit(X_train, y_train)

selected = [feature_cols[i] for i, flag in enumerate(rfe.support_) if flag]

print("\nRFE選択特徴量:", selected)

ranking = pd.Series(rfe.ranking_, index=feature_cols).sort_values()
print("\nRFEランキング")
print(ranking)

plt.figure(figsize=(10, 4))
ranking.plot(kind="bar")
plt.title("RFE Ranking (1 = selected)")
plt.ylabel("Rank")
plt.grid(True, axis="y")
plt.show()


# =========================================
# 15. RFE後 Random Forest
# =========================================
X_train_r = X_train[selected].copy()
X_test_r = X_test[selected].copy()

rf_rfe_model = RandomForestRegressor(
    n_estimators=500,
    random_state=1,
    n_jobs=-1
)
rf_rfe_model.fit(X_train_r, y_train)

y_train_pred_rfe = rf_rfe_model.predict(X_train_r)
y_test_pred_rfe = rf_rfe_model.predict(X_test_r)

rf_rfe_train_metrics = evaluate(y_train, y_train_pred_rfe, "Random Forest (RFE) Train")
rf_rfe_test_metrics = evaluate(y_test, y_test_pred_rfe, "Random Forest (RFE) Test")


# =========================================
# 16. RFEモデルの図
# =========================================
plot_compare(
    date_train, y_train.values, y_train_pred_rfe,
    title="Random Forest (RFE) - Train Fit",
    ylabel="Differenced Price",
    actual_label="Train Actual",
    pred_label="Train Predicted"
)

plot_compare(
    date_test, y_test.values, y_test_pred_rfe,
    title="Random Forest (RFE) - Test Prediction",
    ylabel="Differenced Price",
    actual_label="Test Actual",
    pred_label="Test Predicted"
)

plot_scatter_actual_pred(
    y_test.values,
    y_test_pred_rfe,
    title="Random Forest (RFE) - Actual vs Predicted"
)

rfe_residuals = y_test.values - y_test_pred_rfe

plot_residuals(
    date_test,
    rfe_residuals,
    title="Random Forest (RFE) - Residuals Over Time"
)

plot_hist(
    rfe_residuals,
    title="Random Forest (RFE) - Residual Histogram",
    xlabel="Residual"
)

plot_line(
    date_test,
    np.abs(rfe_residuals),
    title="Random Forest (RFE) - Absolute Error Over Time",
    ylabel="Absolute Error"
)


# =========================================
# 17. RF と RF(RFE) の比較
# =========================================
plt.figure(figsize=(12, 5))
plt.plot(date_test, y_test.values, label="Actual")
plt.plot(date_test, y_test_pred_rf, label="Random Forest")
plt.plot(date_test, y_test_pred_rfe, label="Random Forest (RFE)")
plt.title("Prediction Comparison")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(date_test, y_test.values - y_test_pred_rf, label="Residual: RF")
plt.plot(date_test, y_test.values - y_test_pred_rfe, label="Residual: RF (RFE)")
plt.axhline(0, linestyle="--")
plt.title("Residual Comparison")
plt.xlabel("Date")
plt.ylabel("Residual")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 18. ARモデル
# =========================================
def fit_plot_ar(y_series, dates_series, lags):
    ar_model = AutoReg(y_series, lags=lags, old_names=False)
    ar_result = ar_model.fit()

    print(f"\n[AR({lags})]")
    print(ar_result.summary())

    fitted = ar_result.fittedvalues
    fitted_dates = dates_series.iloc[len(dates_series) - len(fitted):]

    plt.figure(figsize=(12, 4))
    plt.plot(dates_series, y_series, label="Actual")
    plt.plot(fitted_dates, fitted, label=f"AR({lags}) Fitted")
    plt.title(f"AR({lags}) - Fitted vs Actual")
    plt.xlabel("Date")
    plt.ylabel("Differenced Price")
    plt.legend()
    plt.grid(True)
    plt.show()

    residuals = y_series.iloc[len(y_series) - len(fitted):].values - fitted.values

    plt.figure(figsize=(8, 4))
    plt.hist(residuals, bins=20)
    plt.title(f"AR({lags}) - Residual Histogram")
    plt.xlabel("Residual")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

    return ar_result


ar1_result = fit_plot_ar(y, dates, 1)
ar2_result = fit_plot_ar(y, dates, 2)
ar12_result = fit_plot_ar(y, dates, 12)


# =========================================
# 19. MA(1)モデル
# =========================================
ma_model = ARIMA(y, order=(0, 0, 1))
ma_result = ma_model.fit()

print("\n[MA(1)]")
print(ma_result.summary())

ma_fitted = ma_result.fittedvalues

plt.figure(figsize=(12, 4))
plt.plot(dates, y, label="Actual")
plt.plot(dates, ma_fitted, label="MA(1) Fitted")
plt.title("MA(1) - Fitted vs Actual")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()

ma_residuals_full = y.values - ma_fitted.values

plt.figure(figsize=(8, 4))
plt.hist(ma_residuals_full, bins=20)
plt.title("MA(1) - Residual Histogram")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()


# =========================================
# 20. AR / MA テスト予測
# =========================================
y_train_ts = y.iloc[:split]
y_test_ts = y.iloc[split:]
date_test_ts = dates.iloc[split:]

ar1_forecast = rolling_forecast_ar(y_train_ts, y_test_ts, lags=1)
ar2_forecast = rolling_forecast_ar(y_train_ts, y_test_ts, lags=2)
ar12_forecast = rolling_forecast_ar(y_train_ts, y_test_ts, lags=12)
ma_forecast = rolling_forecast_arima(y_train_ts, y_test_ts, order=(0, 0, 1))

ar1_metrics = evaluate(y_test_ts, ar1_forecast, "AR(1) Test")
ar2_metrics = evaluate(y_test_ts, ar2_forecast, "AR(2) Test")
ar12_metrics = evaluate(y_test_ts, ar12_forecast, "AR(12) Test")
ma1_metrics = evaluate(y_test_ts, ma_forecast, "MA(1) Test")

plt.figure(figsize=(12, 5))
plt.plot(date_test_ts, y_test_ts.values, label="Actual")
plt.plot(date_test_ts, ar1_forecast, label="AR(1)")
plt.plot(date_test_ts, ar2_forecast, label="AR(2)")
plt.plot(date_test_ts, ar12_forecast, label="AR(12)")
plt.plot(date_test_ts, ma_forecast, label="MA(1)")
plt.title("Classical Time Series Model Comparison")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 21. 指標一覧（途中）
# =========================================
metrics_df_basic = pd.DataFrame({
    "RF_Test": rf_test_metrics,
    "RF_RFE_Test": rf_rfe_test_metrics,
    "AR1_Test": ar1_metrics,
    "AR2_Test": ar2_metrics,
    "AR12_Test": ar12_metrics,
    "MA1_Test": ma1_metrics
}).T

print("\n=== モデル別テスト評価（途中） ===")
print(metrics_df_basic.sort_values("RMSE"))


# =========================================
# 22. ARMA example
# =========================================
arma_model_full = ARIMA(y, order=(2, 0, 1))
arma_result_full = arma_model_full.fit()

print("\n[ARMA(2,0,1) Full]")
print(arma_result_full.summary())

arma_fitted = arma_result_full.fittedvalues

plt.figure(figsize=(12, 4))
plt.plot(dates, y, label="Actual")
plt.plot(dates, arma_fitted, label="ARMA(2,0,1) Fitted")
plt.title("ARMA(2,0,1) - Fitted vs Actual")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()

arma_residuals_full = y.values - arma_fitted.values

plt.figure(figsize=(8, 4))
plt.hist(arma_residuals_full, bins=20)
plt.title("ARMA(2,0,1) - Residual Histogram")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

arma_one_step = arma_result_full.predict(start=len(y), end=len(y))
print("ARMA one-step prediction:")
print(arma_one_step)


# =========================================
# 23. SARIMA example
# =========================================
sarima_model_full = SARIMAX(
    y,
    order=(1, 1, 1),
    seasonal_order=(0, 0, 0, 0),
    enforce_stationarity=False,
    enforce_invertibility=False
)
sarima_result_full = sarima_model_full.fit(disp=False)

print("\n[SARIMA(1,1,1)x(0,0,0,0) Full]")
print(sarima_result_full.summary())

sarima_fitted = sarima_result_full.fittedvalues

plt.figure(figsize=(12, 4))
plt.plot(dates, y, label="Actual")
plt.plot(dates, sarima_fitted, label="SARIMA Fitted")
plt.title("SARIMA - Fitted vs Actual")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()

sarima_residuals_full = y.values - sarima_fitted.values

plt.figure(figsize=(8, 4))
plt.hist(sarima_residuals_full, bins=20)
plt.title("SARIMA - Residual Histogram")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

sarima_one_step = sarima_result_full.predict(start=len(y), end=len(y))
print("SARIMA one-step prediction:")
print(sarima_one_step)


# =========================================
# 24. ARMA / SARIMA テスト予測
# =========================================
arma_forecast = rolling_forecast_arima(y_train_ts, y_test_ts, order=(2, 0, 1))
sarima_forecast = rolling_forecast_sarima(
    y_train_ts, y_test_ts,
    order=(1, 1, 1),
    seasonal_order=(0, 0, 0, 0)
)

arma_metrics = evaluate(y_test_ts, arma_forecast, "ARMA(2,0,1) Test")
sarima_metrics = evaluate(y_test_ts, sarima_forecast, "SARIMA(1,1,1) Test")

plt.figure(figsize=(12, 5))
plt.plot(date_test_ts, y_test_ts.values, label="Actual")
plt.plot(date_test_ts, arma_forecast, label="ARMA(2,0,1)")
plt.plot(date_test_ts, sarima_forecast, label="SARIMA(1,1,1)")
plt.title("ARMA / SARIMA Test Forecast")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 25. VAR 用データ作成（特異行列対策済み）
# =========================================
# 同じ系列のラグ列を並べず、季節説明変数と組み合わせる
var_cols = ["diff", "month_sin", "month_cos"]
var_df = df[var_cols].dropna().copy()
var_dates = df.loc[var_df.index, "Date"].copy()

var_df = var_df.reset_index(drop=True)
var_dates = var_dates.reset_index(drop=True)

print("\nVAR用データ先頭:")
print(var_df.head())
print("VAR用データ shape:", var_df.shape)

print("\nVAR用 各列の分散")
print(var_df.var())

print("\nVAR用 相関")
print(var_df.corr())


# =========================================
# 26. VAR フル学習
# =========================================
var_result_full, used_lag_full = fit_var_safely(var_df, maxlags=2)

print(f"\n[VAR Full] used_lag={used_lag_full}")
try:
    print(var_result_full.summary())
except Exception as e:
    print("VAR summary() は出せませんでした:", e)

lag_order_full = var_result_full.k_ar
var_fitted = var_result_full.fittedvalues

var_fitted_dates = var_dates.iloc[lag_order_full:].reset_index(drop=True)
var_actual_diff = var_df["diff"].iloc[lag_order_full:].reset_index(drop=True)
var_fitted_diff = var_fitted["diff"].reset_index(drop=True)

plt.figure(figsize=(12, 4))
plt.plot(var_fitted_dates, var_actual_diff, label="Actual diff")
plt.plot(var_fitted_dates, var_fitted_diff, label="VAR Fitted diff")
plt.title("VAR - Fitted vs Actual (diff)")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()

var_residuals_full = var_actual_diff.values - var_fitted_diff.values

plt.figure(figsize=(8, 4))
plt.hist(var_residuals_full, bins=20)
plt.title("VAR - Residual Histogram (diff)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

var_one_step = var_result_full.forecast(var_df.values[-lag_order_full:], steps=1)
print("VAR one-step forecast:")
print(var_one_step)


# =========================================
# 27. VAR テスト予測
# =========================================
var_split = int(len(var_df) * 0.8)
var_train = var_df.iloc[:var_split].copy()
var_test = var_df.iloc[var_split:].copy()
var_test_dates = var_dates.iloc[var_split:].copy().reset_index(drop=True)

var_forecast_df = rolling_forecast_var(var_train, var_test, maxlags=2)

var_test_actual_diff = var_test["diff"].values
var_test_pred_diff = var_forecast_df["diff"].values

var_metrics = evaluate(var_test_actual_diff, var_test_pred_diff, "VAR Test")

plot_compare(
    var_test_dates,
    var_test_actual_diff,
    var_test_pred_diff,
    title="VAR - Test Prediction (diff)",
    ylabel="Differenced Price",
    actual_label="Test Actual",
    pred_label="VAR Predicted"
)

plot_scatter_actual_pred(
    var_test_actual_diff,
    var_test_pred_diff,
    title="VAR - Actual vs Predicted (diff)"
)

var_test_residuals = var_test_actual_diff - var_test_pred_diff

plot_residuals(
    var_test_dates,
    var_test_residuals,
    title="VAR - Residuals Over Time"
)

plot_hist(
    var_test_residuals,
    title="VAR - Residual Histogram",
    xlabel="Residual"
)

plot_line(
    var_test_dates,
    np.abs(var_test_residuals),
    title="VAR - Absolute Error Over Time",
    ylabel="Absolute Error"
)


# =========================================
# 28. 全モデル予測比較
# =========================================
comparison = pd.DataFrame({
    "Actual": y_test.values,
    "RF": y_test_pred_rf,
    "RF_RFE": y_test_pred_rfe,
    "AR1": ar1_forecast,
    "AR2": ar2_forecast,
    "AR12": ar12_forecast,
    "MA1": ma_forecast,
    "ARMA201": arma_forecast,
    "SARIMA111": sarima_forecast
}, index=date_test.values)

print("\n=== テスト予測比較 先頭5行 ===")
print(comparison.head())

plt.figure(figsize=(12, 5))
plt.plot(date_test, np.abs(y_test.values - y_test_pred_rf), label="RF")
plt.plot(date_test, np.abs(y_test.values - y_test_pred_rfe), label="RF_RFE")
plt.plot(date_test, np.abs(y_test.values - ar1_forecast), label="AR1")
plt.plot(date_test, np.abs(y_test.values - ar2_forecast), label="AR2")
plt.plot(date_test, np.abs(y_test.values - ar12_forecast), label="AR12")
plt.plot(date_test, np.abs(y_test.values - ma_forecast), label="MA1")
plt.plot(date_test, np.abs(y_test.values - arma_forecast), label="ARMA(2,0,1)")
plt.plot(date_test, np.abs(y_test.values - sarima_forecast), label="SARIMA(1,1,1)")
plt.title("Absolute Error Comparison")
plt.xlabel("Date")
plt.ylabel("Absolute Error")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 5))
plt.plot(date_test, y_test.values, label="Actual")
plt.plot(date_test, y_test_pred_rf, label="RF")
plt.plot(date_test, y_test_pred_rfe, label="RF_RFE")
plt.plot(date_test, ar1_forecast, label="AR1")
plt.plot(date_test, ar2_forecast, label="AR2")
plt.plot(date_test, ar12_forecast, label="AR12")
plt.plot(date_test, ma_forecast, label="MA1")
plt.plot(date_test, arma_forecast, label="ARMA(2,0,1)")
plt.plot(date_test, sarima_forecast, label="SARIMA(1,1,1)")
plt.title("Model Forecast Comparison")
plt.xlabel("Date")
plt.ylabel("Differenced Price")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 29. 指標一覧
# =========================================
metrics_df = pd.DataFrame({
    "RF_Test": rf_test_metrics,
    "RF_RFE_Test": rf_rfe_test_metrics,
    "AR1_Test": ar1_metrics,
    "AR2_Test": ar2_metrics,
    "AR12_Test": ar12_metrics,
    "MA1_Test": ma1_metrics,
    "ARMA201_Test": arma_metrics,
    "SARIMA111_Test": sarima_metrics,
    "VAR_Test": var_metrics
}).T

print("\n=== モデル別テスト評価 ===")
print(metrics_df.sort_values("RMSE"))


# =========================================
# 30. VAR を別枠で比較表示
# =========================================
var_comparison = pd.DataFrame({
    "Actual_diff": var_test_actual_diff,
    "VAR_diff_pred": var_test_pred_diff,
    "AbsError": np.abs(var_test_actual_diff - var_test_pred_diff)
}, index=var_test_dates.values)

print("\n=== VAR テスト予測比較 先頭5行 ===")
print(var_comparison.head())

```
